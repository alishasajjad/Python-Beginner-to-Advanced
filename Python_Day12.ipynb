### Series

```python
# A Series is one column -> only one column

# Series – A one-dimensional labeled array
import pandas as pd

s = pd.Series([10, 20, 30])
print(s)

## In Output -> 0,1,2 are indexes
```

### Access Value

```python
s = pd.Series([10, 20, 30])
print(s[0])
```

### Custom Index

```python
## Custom Index

# Instead of numbers, -> give names

a = [12,23,45]
s = pd.Series(a,index=["A","B","C"])
print(s)
```

### Access

```python
print(s["B"])
```

### Series from Dictionary

```python
## Dictionary

marks = {
    "Ali" : 90,
    "Alisha" : 100
}

## Convert to Series

s = pd.Series(marks)
print(s)

# Dictionary keys become indexes
```

```python
## Select Specific Keys

s1 = pd.Series(marks,index=["Ali"])
print(s1)
```

## DataFrame

```python
# DataFrame – A two-dimensional table with rows and columns
## A DataFrame is a complete table

import pandas as pd

## Create DataFrame
data = {
    "Name": ["Alisha", "Ali", "Zarmeen"],
    "Age": [25, 30, 20],
    "City": ["Lahore", "Sahiwal", "Multan"]
}

df = pd.DataFrame(data)
print(df)
```

### DataFrame from Dictionary

```python
## DataFrame from Dictionary

data = {
    "Calories" : [420,340,201],
    "Duration": [40,50,45]
}

df = pd.DataFrame(data)
print(df)

## Each key becomes a column
```

### loc[]

```python
## loc[]

## Used to access rows

print(df.loc[0]) # one row in output
```

```python
## Multiple Rows

print(df.loc[[0,1]])

## Notice the double brackets [[ ]] because ->  passing a list of row indexes
```

### Named Index

```python
data = {
    "Calories" : [420,340,201],
    "Duration": [40,50,45]
}

df = pd.DataFrame(data,index=["day1","day2","day3"])
print(df)
```

### Access

```python
print(df.loc["day2"])
```
