## Normal Distribution

```python
## Normal Distribution

# Also called Gaussian Distribution


#It Looks like :

#           /\
#         /    \
#       /        \
#_____/___________\____

# This is called Bell Curve
```

### Generate Normal Distribution

```python
## Generate Normal Distribution

from numpy import random
x = random.normal(size=(2,3))
print(x)
```

### Parameters

```python
# It has three parameters:

# loc - (Mean) where the peak of the bell exists -> Center of Graph

# scale - (Standard Deviation) how flat the graph distribution should be -> less flat means -> small scale , ---- -> more flat -> means -> large scale

# size - Shape -> The shape of the returned array

# size=(2,3) -> Creates -> 2 rows and 3 columns
```

### Visualization

```python
sns.displot(random.normal(size=1000), kind='kde')

## Produces a bell curve

## The curve of a Normal Distribution is also known as the Bell Curve because of the bell-shaped curve
```

```python
# Example
from numpy import random
x = random.normal(loc=1, scale=2, size=(2, 3))
print(x)
```

## Binomial Distribution

```python
## Binomial Distribution

# Used when -> There are only 2 possible outcomes
# It describes the outcome of binary scenarios, e.g. toss of a coin, it will either be head or tails
```

### Parameters

```python
## Parameters

## It has three parameters:

# n - number of trials e.g -> 10 tosses -> n = 10

# p - probability of occurrence of each trial (e.g. for toss of a coin 0.5 each)

# size - The shape of the returned array
```

```python
random.binomial(n=10,p=0.5,size=10)
```

```python
## Visualization
sns.displot(random.binomial(n=10,p=0.5,size=1000))
```

### Normal vs Binomial

```python
# Normal -> Continuous
170.11
170.12
170.13
# Infinite values

# Binomial -> Discrete
# Only
0
1
2
3
# No decimal outcomes
```

```python
data = {
    "normal" : random.normal(loc=50,scale=5,size=1000),
    "binomial": random.binomial(n=100,p=0.5,size=1000)
}

sns.displot(data, kind='kde')
plt.show()
```

## Poisson Distribution

```python
## Poisson Distribution

# Poisson Distribution is a Discrete Distribution

# Used to estimate -> How many times an event happens

## Examples
# Calls received per hour
# Customers entering a shop
# Emails received today
# Accidents per month

```

### Parameters

```python
## It has two parameters:

# lam - Average occurrence ->  rate or known number of occurrences e.g. 2 for above problem

# size - The shape of the returned array
```

```python
random.poisson(lam=2,size=10)
```

```python
## Visualization
sns.displot(random.poisson(lam=2,size=1000))
```

### Binomial vs Poisson

```python
## Binomial
# Fixed number of trials
# Example -> 10 tosses

## Poisson
# No fixed trials
# Only average occurrence
# Example -> Customers arriving every hour
```

```python
data = {
    "normal" : random.normal(loc=50,scale=7,size=1000),
    "poisson" : random.poisson(lam=50, size=1000)
}

sns.displot(data, kind= 'kde')
```

### Difference Between Binomial and Poisson Distribution

```python
# Binomial distribution only has two possible outcomes, whereas poisson distribution can have unlimited possible outcomes

# But for very large n and near-zero p binomial distribution is near identical to poisson distribution such that n * p is nearly equal to lam
```

```python
data = {
    "binomial" : random.binomial(n=1000, p=0.01,size=1000),
    "poisson" : random.poisson(lam=10, size=1000)
}

sns.displot(data, kind= 'kde')
```

## Uniform Distribution

```python
## Uniform Distribution

# Used to describe probability where every event has equal chances of occurring

# e.g. Generation of random numbers


#Every value has -> Equal chance
```

### Parameters

```python
## It has three parameters:

# low - lower bound - default 0.0

# high - upper bound - default 1.0

# size - The shape of the returned array
```

```python
random.uniform(size=(2,3))
```

```python
## Visualization

sns.displot(random.uniform(size=1000), kind='kde')
```

## Logistic Distribution

```python
## Logistic Distribution
# Logistic Distribution is used to describe growth

# Used mainly in -> Logistic Regression, Neural Networks, Deep Learning

# Looks very similar to Normal Distribution
```

### Parameters

```python
## loc - mean, where the peak is ->  Default 0

# scale - standard deviation, the flatness of distribution -> Default 1

# size - The shape of the returned array
```

```python
random.logistic(size=100)
```

```python
## Visualization of Logistic Distribution
sns.displot(random.logistic(size=1000), kind='kde')
```

```python
random.logistic(loc=1,scale=2,size=(2,3))
```

### Difference Between Logistic and Normal Distribution

```python
# Both distributions are near identical, but logistic distribution has more area under the tails, meaning it represents more possibility of occurrence of an event further away from mean

# For higher value of scale (standard deviation) the normal and logistic distributions are near identical apart from the peak
```

```python
data = {
    "noraml" : random.normal(scale=2, size=1000),
    "logistic" : random.logistic(size=1000)
}
sns.displot(data, kind='kde')
```

```python
# Normal Distribution → Height, Weight, IQ, Exam Marks
# Binomial Distribution → Coin Toss, Pass/Fail, Yes/No
# Poisson Distribution → Customers per hour, Calls per minute, Accidents per day
# Uniform Distribution → Random numbers where every value is equally likely
# Logistic Distribution → Logistic Regression and Neural Networks
```
