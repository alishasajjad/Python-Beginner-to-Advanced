# Python Range

```python
###Python Range 

#range() -> Python built-in function -> make numbers sequence 
#mostly use in for loop

for i in range(5):
    print(i)

#range is not a list
# range -> object

x = range(10)
print(x)

print(list(x)) #list

##Syntax  -> 3 parameters
#not imp to write three
#range(start, stop, step)
#start include, end exclude
#1. range(stop)

range(5) #-> Start = 0, Stop = 5, Step = 1
for i in range(5):
    print(i)

#2. range(start, stop)
for i in range(3,8):  #Start = 3, Stop = 8 step = 1 
    print(i)

#3. range(start, stop, step) #step means -> jump step 
for i in range(2,11,2):
    print(i)

for i in range(5,15,5):
    print(i)

###Negative STep -> reverse counting
#for i in range(0,10,-2):  #Nothing print
#for i in range(10,0,2):  #Nothing print
for i in range(10,0,-2):
    print(i)

#Range object 
r = range(5) #range(0, 5)
print(r)
print(type(r)) #range

#Convert into list 
print(list(range(5)))
print(list(range(2,8,3)))
#print(list(range(2,500,0))) #error

##Range Indexing 
# You can access elements of a range object using indexes, just like a list
r = range(10)
print(r[2]) 

##Range Slicing
r = range(10)
print(r[:2]) 

#if list see
print(list(r[:2]))

r = range(20)
print(list(r[5:10]))

###MemberShip Testing
# in operate check value is in range or not 
r = range(0,12,2)
print(6 in r)
print(7 in r)

##len()
r = range(0,10,2)
print(len(r))

## Range Immutable -> means -> not changeable 
r = range(5)
#r[0] = 10 #typError

#Range vs List
#   Range	                  List
#  Immutable	            Mutable
#Less memory use     	More memory use
#for Mostly loops       for Data store
#Type = range	            Type = list

# range() creates a sequence of numbers.
# Syntax: range(start, stop, step).
# If only stop is provided, start = 0 and step = 1 by default.
# Start is inclusive (included in the sequence).
# Stop is exclusive (not included in the sequence).
# Step controls the increment or decrement.
# range is an immutable object, not a list.
# Use list(range(...)) to view its values as a list.
# Use len(range(...)) to get the total number of elements.
# The 'in' operator can be used to check whether a number exists in the range or not.
```

# Python Arrays

```python
###Python Arrays
#In Python there is no built-in arrays like c++ or java 
#Python normally use list as an array like

#for real arrays python use Numpy Library
#import numpy as np

#variable which store multiple values
#Without Array

car1 = "Ford"
car2 = "BMW"
car3 = "Honda"

#With Array (List)
cars = ["Plane","BMW","Civic"]
# index   0       1      2

#syntax 
# array_name = [value1, value2, value3]
cars = ["Plane","BMW","Civic"]
#print array
print(cars)

#array elements access through index
print(cars[0])

##Negative Indexing
print(cars[-1]) #last item

##Value change in Array 
cars = ["Plane","BMW","Civic", "Anywhere Door"]
cars[0] = "Ship"
print(cars)

###Array Length ->items quantity
print(len(cars))

##Loop through Array
cars = ["Plane","BMW","Civic", "Anywhere Door"]

for x in cars:
    print(x)

###append() -> add new element in last
#syntax
# array.append(value)
cars = ["Plane","BMW","Civic", "Anywhere Door"]
cars.append("Honda")
print(cars)

#pop() -> remove through index
#array.pop(index)
cars = ["Plane","BMW","Civic", "Anywhere Door"]
cars.pop(1)
print(cars)

#cars.pop() # if index not -> last element delete

##remove() -> delete through value
cars = ["Plane","BMW","Civic", "Anywhere Door"]
cars.remove("Plane") #if any value which is not exist in array then ->valueError
print(cars)

#insert() -> add an item at specific position
cars = ["Plane","BMW","Civic", "Anywhere Door"]
cars.insert(0,"alisha")
print(cars)

##extend() -> join 2 lists
cars = ["Plane","BMW","Civic", "Anywhere Door"]
names = ["alisha", "ali"]
cars.extend(names)
print(cars)

###copy()
cars = ["Plane", "Anywhere Door"]
new1 = cars.copy()
print(new1)

#clear()
# all items delete
cars = ["Plane","BMW","Civic", "Anywhere Door"]
cars.clear()
print(cars)

##count() -> how many times a value exist
cars = ["Plane","BMW","Civic", "Anywhere Door", "Plane"]
print(cars.count("Plane"))

#Index -> value in which position 
cars = ["Plane","BMW","Civic", "Anywhere Door"]
print(cars.index("Plane"))

##reverse()
cars = ["Plane","BMW","Civic", "Anywhere Door"]
cars.reverse() #order reverse 
print(cars)

#sort() ->ascending order
cars = ["Plane","BMW","Civic", "Anywhere Door"]
cars.sort(reverse=True) #descending order , if cars.sort() -> ascending order
print(cars)

##Complete Example
cars = ["Plane","BMW","Civic", "Anywhere Door"]
print(cars)

cars.append("Civic")

cars.remove("BMW")

cars.insert(1,"Audi")

print(cars)

for car in cars:
    print(car)

print(len(cars))

### Method              Purpose
# append(x)             Adds a value at the end of the list.
# insert(i, x)          Adds a value at a specific position.
# extend(list)          Joins two lists together.
# remove(x)             Removes the specified value.
# pop(i)                Removes the element at the given index.
# clear()               Removes all elements from the list.
# copy()                Creates a copy of the list.
# count(x)              Returns how many times a value appears.
# index(x)              Returns the index of a value.
# sort()                Sorts the list in ascending order.
# reverse()             Reverses the order of the list.
# len(list)             Returns the total number of elements.
```

# Python Iterator

```python
## Python Iterator

# Iterator means -> an object that contains multiple values and returns them one by one.

# list, tuple, string, set, dictionary, range

student1 = ["alisha", "sajjad", "aiman", "sajjad"]
# we run a loop on it
for x in student1:
    print(x)

# An iterator is an object that accesses the elements of an iterable one at a time.

#list
student1 = ["alisha", "sajjad", "aiman", "sajjad"]
# Iterator works like this:
# alisha -> sajjad -> aiman -> sajjad
# One value at a time

### Iterable and Iterator

# iterable -> store data -> list, set, tuple, dictionary, range, string
student1 = ["alisha", "sajjad", "aiman", "sajjad"]

# Iterator -> converts an iterable into a sequence of values
#iter() :  alisha -> sajjad -> aiman -> sajjad ->end

### iter() function -> used to create an iterator
student1 = ["alisha", "sajjad", "aiman", "sajjad"]
x = iter(student1)

# student1 -> iterable (list)
# x -> iterator

## next() -> used to get the next value from the iterator
#next() 
student1 = ["alisha", "sajjad", "aiman", "sajjad"]
x = iter(student1)  # Converts the iterable into an iterator
print(next(x))  # Used to get values
print(next(x))
print(next(x))
print(next(x))
# print(next(x))      # Error -> StopIteration -> no values left in the iterator
# The iterator has no more values.

### Iterable STring
name = "alisha"
y = iter(name)
print(next(y))
print(next(y))
print(next(y))
print(next(y))

## Tuple Iterator 
t = ("alisha", "sajjad", "ibrahim", "asif")
z = iter(t)
print(next(z))
print(next(z))

### for loop uses an iterator -> Important
student1 = ["alisha", "sajjad", "aiman", "sajjad"]

for x in student1:
    print(x)

# The for loop does not read the list directly.
# Python does this internally, which is why the for loop uses an iterator.
s = iter(student1)
print(next(s))
print(next(s))
print(next(s))

### Internally 
student1 = ["alisha23", "sajjad34", "aiman43", "sajjad32"]
for x in student1:
    print(x)

# python internally does this:
it = iter(student1)
while True:
    try:
        x = next(it)
        print(x)
    except StopIteration:
        break

# Because of this, the for loop never becomes infinite.

### Create Your Own Iterator
# We can also make a class behave like an iterator using two methods. 

#1. __iter__()
#2. __next__()

#1. iter() -> starts the iterator

#def __iter__(self): 
# Initial values are set here

#2. next() -> returns the next value every time it is called

class MyNumber:
    def __iter__(self):
        self.x = 1
        return self
    
    def __next__(self):
        y = self.x
        self.x += 1
        return y 
#never stop -> infinite iterator
### 
obj = MyNumber()

it = iter(obj)

print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))

#To stop an iterator -> use raise StopIteration

class MyNumbers:
    
    def __iter__(self):
        self.x = 1
        return self
    def __next__(self):
        
        if self.x <=5: #When the condition becomes False, the iteration stops
            y = self.x
            self.x +=1
            return y
        else:
            raise StopIteration
    

# Function          Work
# iter()            Converts an iterable into an iterator.
# next()            Returns the next value.
# for               Internally uses iter() and next().
# StopIteration     Stops the iterator.

# To create a custom iterator, implement the __iter__() and __next__() methods in a class


```

```python
##Python DateTime

#Current Date, Current Time ,Age Calculate, Birthday, Login Time, Attendance, File Date, Clock

#in python, there is no built-in data type for date and time
# for this python make a module 
# datetime ->handle time and date

import datetime #python load module dateTime

#current date and time

x = datetime.datetime.now() #current system datetime store in variable 
print(x)

#Diagram
#2026-07-07 09:01:25.004794
#Year
#   Month
#      Day
#         Hour
#            Minute
#                Second
#                       Microsecond

#if only year print
x = datetime.datetime.now()

print(x.year) #year print from x which store datetime

#if only month,day,hour,minutes,seconds,microsecond
print(x.month)
print(x.day)
print(x.hour)
print(x.minute)
print(x.second)
print(x.microsecond)

###Current WeekDay
import datetime

x = datetime.datetime.now()
print(x.strftime("%A")) #day

# strftime() -> Converts a date and time into the desired format

import datetime

x = datetime.datetime.now()
print(x.strftime("%B")) #month

##date object 
#sometimes we not need current date 
#need format date -> any
#Birthday
import datetime

birthday = datetime.datetime(2006, 8, 8)  #time not write -> python automatically write time 00:00:00
print(birthday) 

#time write

x = datetime.datetime(2009,9,30,10,13,34) #if a datetime is write which not exist in real -> wrong datetime, -> valueerror
print(x)

###strftime() Formatting ->imp function
#syntax
#date.strftime("format")

#%Y -> Full Year
print(x.strftime("%Y")) #2026

#%y -> last 2 digit year 
print(x.strftime("%y")) #26

#%B -> Full Month name
print(x.strftime("%B")) #July

#%b -> Short Month name 
print(x.strftime("%b")) #Jul

#%A -> full day name
print(x.strftime("%A")) #Wednesday

#%a -> short day name
print(x.strftime("%a")) #Wed

#%m -> month number
print(x.strftime("%m")) #07

#%H -> 24 Hour format
print(x.strftime("%H")) # 10

#%I -> 12 hour format 
print(x.strftime("%I")) #23 -> 11

#%p -> AM, PM
print(x.strftime("%p")) #AM

#%M -> Minute 
print(x.strftime("%M")) #13

#%S -> Second
print(x.strftime("%S")) #25

#%f -> Microsecond
print(x.strftime("%f")) #000000

###Some Useful Format

#Format 1 
print(x.strftime("%d/%m/%Y"))  #12/12/2009

#Format 2
print(x.strftime("%d-%m-%Y"))  #12-12-2009

#Format 3
print(x.strftime("%B %m, %Y"))   #July 07, 2026

#Format 4
print(x.strftime("%A")) #day

#Format 5
print(x.strftime("%A %d %B %Y")) #Tuesday 07 July 2026

#Format 6
print(x.strftime("%I:%M %p")) #09:09 PM


####Student Login Time

import datetime

login = datetime.datetime.now()

print("login time:",  login)

##Attendance System 
import datetime

today = datetime.datetime.now()
print(today.strftime("%d/%m/%Y"))

###Digital Clock
import datetime

clock = datetime.datetime.now()
print(clock.strftime("%I:%M:%S.%f %p"))

####Birthday Object
import datetime

birthday = datetime.datetime(2005,1,8)
print(birthday.strftime("%a"))

###
#Format	   Meaning	        Example Output
#%Y	      Full Year	           2026
#%y	      2-digit Year	        26
#%m	      Month Number	        07
#%B	      Full Month Name	   July
#%b	      Short Month Name	    Jul
#%d	      Day	                07
#%A	      Full Weekday	       Tuesday
#%a	      Short Weekday	        Tue
#%H	      Hour (24-hour)	    09
#%I	      Hour (12-hour)	    09
#%M	      Minute	            01
#%S	      Second	            25
#%p	      AM/PM	                AM

##
#import datetime
#        ↓
#datetime.datetime.now()
#        ↓
#Current Date & Time

#datetime.datetime(Year, Month, Day)
#        ↓
#Custom Date

#strftime()
#        ↓
#Date convert to required format
```
