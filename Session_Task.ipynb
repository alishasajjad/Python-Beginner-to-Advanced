```python
## Currency Converter (USD to PKR)

rate = 280
usd = float(input("Enter Amount in USD: "))
pkr = usd * rate

print(f"{usd} USD = {pkr:.2f} PKR")
```

```python
## BMI Calculator

weight = float(input("Enter Weight (kg): "))
height = float(input("Enter Height (m): "))

bmi = weight / (height ** 2)

print(f"\n BMI: {bmi:.2f}")

if bmi < 18.5:
    print("Underweight")
elif bmi < 25:
    print("Normalweight")
elif bmi < 30:
    print("Overweight")
else:
    print("Obese")

```

```python
## SHopping LIST

shopping_List = []

while True:

    choice = input("Choice (1-6): ")

    if choice == "1":
        item = input("Enter item: ")
        shopping_List.append(item)
        print("Item added")

    elif choice == "2":
        item = input("Enter Item for remove: ")

        if item in shopping_List:
            shopping_List.remove(item)
            print("Removed")
        else:
            print("Not Found")

    elif choice == "3":
        print(shopping_List)

    elif choice == "4":
        item = input("Enter Search: ")

        if item in shopping_List:
            print("Item Found")
        else:
            print("Not Found")

    elif choice == "5":
        shopping_List.sort()
        print(shopping_List)

    elif choice == "6":
        print("Exit")
        break

    else:
        print("Invalid Choice")
```

```python
## Calculator Using Functions

def add(a,b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return  a * b

def divide(a,b):
    return a / b

def calculator():
    a = float(input("Enter Number first: "))
    b = float(input("Enter Number second: "))
    
    choice = input("Select (1-4): ")
    
    if choice == "1":
        print(add(a,b))
    elif choice == "2":
        print(subtract(a,b))
    elif choice == "3":
        print(multiply(a,b))
    elif choice == "4":
        print(divide(a,b))
    else:
        print("Invalid")

calculator()
```
