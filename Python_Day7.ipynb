## 1. Python File Open

```python
#File Handling 

#File Handling means working with files in Python

#Using Python -> Create a file , Open a file, Read data from a file, Write data into a file, Update a file, Delete a file

###The open() Function

#Python uses the open() function to open a file

#Code
f11 = open("student.txt", "r")

data = f11.read()
print(data)

f11.close()

##Syntax 
#f = open("filename", "mode")

#Read
f = open("student.txt", "r")

#"students.txt" → File name
#"r" → Mode (Read)

###File Mode 

#1. Read Mode ("r")
#Used to read a file

f1 = open("student.txt", "r")

# If file exists: Python opens the file successfully
# If the file does not exist -> FileNotFoundError is raised

#Remember: "r" is the default mode.

#2. Write Mode ("w")
f2 = open("student.txt", "w")
f2.write("Hello Sir")

#If the file does not exist, Python creates it
#If the file already exists, all old data is deleted before writing

#3. Append Mode ("a")

f3 = open("student.txt", "a")
f3.write("\nAliSha Sajjad 373")

#The old data is not deleted
#If the file does not exist, Python creates it

#4. Create Mode ("x")

#Used to create a new file only

f4 = open("newFile.txt", "x")

#If the file does not exist → It is created.
#If the file already exists → Python raises -> FileExistsError

###Text Mode ("t")
#Text mode is used for text files

#Examples -> .txt, .py, .csv

f5 = open("student.txt", "rt")

#r = Read
#t = Text

#Since both are defaults, these two statements are the same:
#f = open("students.txt") and f = open("students.txt", "rt")

##Binary Mode ("b")

#Binary mode is used for non-text files such as: Images, Videos, PDFs, Audio files

f6 = open("photo4.jpeg", "rb")

#r = Read
#b = Binary

#Why are "r" and "t" the default?

f7 = open("student.txt") 

#Python automatically assumes:
f8 = open("student.txt", "rt")

#So don't need to write "rt" every time

#Important
f9 = open("student.txt", "r") #-> and the file does not exist, Python gives:FileNotFoundError

f10 = f = open("student.txt", "w")
f11 = open("students.txt", "a")
#Python creates the file automatically if it does not already exist

#Mode	Purpose	    If file does not exist
# "r"	Read	     ❌ Error (FileNotFoundError)
# "w"	Write	     ✅ Creates a new file
# "a"	Append	     ✅ Creates a new file
# "x"	Create	     ❌ Error if file already exists
# "t"	Text mode	 Default mode
# "b"	Binary mode	 Used for images, PDFs, videos, etc.
```

## 2. Python Read Files

```python
#Python Read Files

#Create demoFile.txt
#Write -> 

#Python uses the built-in open() function to open a file

f = open("demo.txt")
print(f.read())
f.close()
#open("demo.txt") opens the file
#f.read() reads the complete content of the file
#print() displays the content on the screen

#Opening a File from Another Location

#If the file is not in the current folder, provide its full path

f1 = open("C:\\Users\\DELL\\Desktop\\Intern\\newLocation.txt")  
print(f1.read())

#Python will open the file from the specified folder

#Using the with Statement
with open("demo.txt") as f2:
    print(f2.read())

#Why use with?
#It automatically closes the file
#No need to write f.close()
#It is safer and easier to use

#Closing a File
#If you do not use with, you should close the file manually
f3 = open("student.txt")
print(f3.readline())
f3.close()

#Why should we close a file?
#It frees system resources
#It saves any pending changes
#It prevents file corruption or resource leaks

###Reading Only Part of a File

#The read() method can take a number as an argument
with open("demo.txt") as f4:
    print(f4.read(8)) 
#read(5) -> Read only the first 8 characters of the file

###Reading One Line
#Use the readline() method

with open("demo.txt") as f5:
    print(f5.readline())

#If file data one three lines then output show only one first line

###Reading two line
#Call readline() twice
with open("demo.txt") as f6:
    print(f6.readline())
    print(f6.readline())

#First call → reads the first line
#Second call → reads the second line


###Reading the File Line by Line
#Use a for loop

with open("demo.txt") as f7:
    for x in f7:
        print(x)

#The loop reads and prints one line at a time until the end of the file

##Difference Between read() and readline()

#read()
print(f.read())
#Reads the entire file.
#Returns all content at once.

#readline()
print(f.readline())
#Reads only one line.
#Each call reads the next line.

#read(5)
print(f.read(5))
#Reads only the first 5 characters

#Method	          Purpose
#open()	        Opens a file
#read()	        Reads the entire file
#read(5)	    Reads the first 5 characters
#readline()   	Reads one line at a time
#with open()	Opens the file and closes it automatically
#close()    	Closes the file manually
```

## 3. Python File Write/Creates

```python
#Writing to a File in Python

#Python allows you to write data into a file

#There are mainly two modes for writing to an existing file:
# "a" → Append
# "w" → Write (Overwrite)

#1. Append Mode ("a")

# Append means add new data at the end of the file

with open("demo.txt", "a") as f1:
    f1.write("Now the file has more content!")

with open("demo.txt") as f:
    print(f.read())

#Opens the file in append mode
#Keeps the old data
#Adds the new text at the end

#Think of it like writing a new sentence at the bottom of a notebook

#2. Write Mode ("w")

# Write mode replaces all existing content in the file
with open("student.txt") as f:
    print(f.read())

with open("student.txt", "w") as f:
    f.write("I have deleted the content!")

with open("student.txt") as f:
    print(f.read())

#Opens the file in write mode
#Deletes all old data
#Writes the new data

#Think of it like erasing everything from a notebook and writing new text

###Create a New File

# Python can also create new files

#There are three ways:

#1. Using "x" Mode
f11 = open("myfile.txt", "x")

#Creates a new file.
#If the file already exists, Python gives an error.

#If myfile.txt already exists: -> FileExistsError

#2. Using "a" Mode

with open("myfile.txt", "a") as f:
    f.write("Hello12")

with open("myfile.txt") as f:
    print(f.read())
    
#If the file exists → Open it.
#If the file does not exist → Create it automatically.

#3. Using "w" Mode

with open("myfile1.txt", "w") as f:
    f.write("Hello_file1")

with open("myfile1.txt") as f:
    print(f.read())

#If the file exists → Delete old content and write new content.
#If the file does not exist → Create a new file.



```

## 4. Python Delete File

```python
#Python Delete File

#Python allows you to delete files and folders using the os module

import os

#The os module provides functions to work with files and directories

#Delete a File

import os
os.remove("demo1.txt") #The file demo1.txt is deleted from your computer

#import os → Imports the operating system module.
#os.remove("demo1.txt") → Deletes the file named demo1.txt.

#If the file exists, it will be deleted permanently

##What if the File Does Not Exist?

#If you try to delete a file that does not exist, Python raises an error

import os
#os.remove("abc.txt")

#Error -> FileNotFoundError: [Errno 2] No such file or directory: 'abc.txt'
#This happens because Python cannot find the file

##Check if a File Exists Before Deleting

import os

if os.path.exists("student1.txt"):
    os.remove("student1.txt")
else:
    print("The file does not exist")


#os.path.exists("student.txt")
#Returns True if the file exists
#Returns False if the file does not exist

#If the file exists, Python deletes it

#Otherwise, it prints: -> The file does not exist

#This is the safer way to delete files

###Delete a Folder

# To delete an empty folder, use os.rmdir()

import os
os.rmdir("myfolder") #The folder myfolder is deleted

#os.rmdir() removes a directory (folder)
#The folder must be empty

##What if the Folder is Not Empty?

#Suppose the folder contains files:

#myfolder1/
#    file1.txt

import os
os.rmdir("myfolder1")

#OSError: Directory not empty

#This happens because os.rmdir() can delete only empty folders.

#Function	                Purpose
#import os	             Imports the OS module
#os.remove("file.txt")	    Deletes a file
#os.path.exists()	     Checks whether a file exists
#os.rmdir("folder")	      Deletes an empty folder
```
