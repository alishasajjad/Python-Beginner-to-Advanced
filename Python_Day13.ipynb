# Pandas - Cleaning Data

## Data Cleaning

```python
## Data Cleaning

# -> Data cleaning means fixing bad data in your data set

# Real-world datasets are never perfect
# They may contain:
##  -> Missing values (NaN)#
##  -> Wrong data format
##  -> Wrong values
##  -> Duplicate rows

# Before applying Machine Learning, we clean the data
```

```python
import pandas as pd

# Example

df = pd.read_csv("students.csv")
print(df)


## problems -> Missing Age, Duplicate Rows, NaN values, Wrong age etc 
```

### Fill only one column

```python
df2 = pd.read_csv("students.csv")
```

```python
df2
```

```python
df2.fillna({"Age": 23}, inplace = True)
```

## Mean

```python
## Mean
# Mean = Average

cgpa_mean = df2["CGPA"].mean()
print(cgpa_mean)
```

### Fill Missing Value in CGPA Column with Mean

```python
df2.fillna({"CGPA" :cgpa_mean}, inplace = True)
```

## Median

```python
## Median

# Median = Middle Value

median_marks = df2["Marks"].median()
print(median_marks)
```

## Mode

```python
## Mode

# Mode -> Most frequent value

age_mode = df2["Age"].mode()
print(age_mode)
```

## Which one should we use?

```python
# Numeric data -> Mean, Median

# Categorical data -> Mode
```

## Wrong Format

```python
df2
```

```python
## Wrong Format

# Sometimes data has incorrect type

# Example
# 2020/12/01
# 2020/12/02
# 20201203

# Last one has different format
```

```python
## Convert Everything to Date 
# Attendance column data -> Convert into date format  (Example)

df2["Attendance"] = pd.to_datetime(df["Attendance"])
```

```python
df2
```

## NaT

```python
# NaT

## If conversion fails -> NaT
## NaT -> means -> Not a Time

# Equivalent of NaN for dates

# Example

# NaN → Missing Number

# NaT → Missing Date
```

## Remove missing dates

```python
df2.dropna(subset = ["Attendance"], inplace = True)
# Only rows with missing Date are removed
```

```python
df2
```

## Wrong Data

```python
## Wrong Data

# Example

# Workout duration

# 30
# 45
# 60
# 450
# 50

# 450 minutes is unrealistic
```

### Replace manually

```python
df2.loc[1, "CGPA"] = 3.44
#      index, column
```

```python
df2
```

## Replace values using rule

```python
for x in df2.index:
    if df2.loc[x, "CGPA"] > 3.45:
        df2.loc[x, "CGPA"] = 3.01
```

```python
df2
```

### Delete Wrong Rows

```python
for x in df2.index:
    if df2.loc[x, "CGPA"] > 3.44:
        df2.drop(x, inplace=True)
```

```python
df2
```

## Duplicate Data

```python
## Find Duplicates 

df2.duplicated()

# False means -> No duplicate
# True means -> DUplicate Data
```

## Remove duplicates

```python
df2.drop_duplicates(inplace=True)
```

## Correlation values

```python
# Correlation values

#+1 -> Perfect Positive

## Example: 
# Study ↑
# Marks ↑
```

```python
# 0

# 0 -> No relation

## Example:

# Shoe Size
# Salary

# No connection
```

```python
# -1

# -1 -> Perfect Negative

## Example

#Price ↑
#Demand ↓

# Opposite direction
```

## General interpretation

```python
## General interpretation

## Correlation	Meaning
#     1	       Perfect Positive
#   0.8–1	   Strong Positive
#   0.6–0.8	   Good Positive
#     0	       No Relation
#  -0.6 to -1  Strong Negative
#     -1	   Perfect Negative


##Example

#Duration vs Calories
# 0.92
# Longer workout → More calories burned
# Good correlation

# Example

# Duration vs Maxpulse
# 0.009
# Almost no relationship
```

# Complete Data Cleaning Workflow

```python
import pandas as pd

# Step 1: Load dataset
df7 = pd.read_csv("students.csv")

# Step 2: Check data
df7.info()

# Step 3: Check missing values
df7.isnull().sum()

# Step 4: Fill or remove missing values
df7.fillna(df.mean(numeric_only=True), inplace=True)

# Step 5: Fix data types
df7["Attendance"] = pd.to_datetime(df["Attendance"], errors="coerce")

# Step 6: Remove rows with invalid dates
df.dropna(subset=["Attendance"], inplace=True)

# Step 7: Remove duplicates
df.drop_duplicates(inplace=True)

# Step 8: Check correlation
print(df.corr(numeric_only=True))

# Step 9: Visualize data
df.plot()
plt.show()
```

```python
# Topic	                   Function
#Remove missing values	    dropna()
#Replace missing values	    fillna()
#Average	                mean()
#Middle value	            median()
#Most frequent value	    mode()
#Convert to date         	pd.to_datetime()
#Remove duplicate rows	    drop_duplicates()
#Find duplicate rows	    duplicated()
#Access/update a cell	    loc[]
#Correlation	            corr()
#Plot graph	                plot()
#Scatter plot	            plot(kind="scatter")
#Histogram	                plot(kind="hist")
```
