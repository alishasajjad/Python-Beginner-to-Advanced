# NumPy ufuncs (Universal Functions)

```python
## What are ufuncs?

## ufunc stands for Universal Function.

## These are built-in NumPy functions that work directly on NumPy arrays (ndarray)
## Instead of processing one element at a time using loops, ufuncs process the entire array at once

```

```python
# Why use ufuncs?

## ufuncs are much faster than Python loops because they use vectorization

#Advantages:

# - Faster execution
# - Less code
# - Works directly with arrays
# - Supports broadcasting
# - Supports methods like reduce(), accumulate()
```

#### Vectorization

```python
## What is Vectorization?

# Vectorization means performing an operation on the entire array at once instead of using loops

## Without Vectorization

x = [1,2,3,4,5]
y = [2,3,789,0,1]

z = []

for i,j in zip(x,y):
    z.append(i+j)

print(z)

## Python adds elements one by one using a loop
```

```python
## With Vectorization (ufunc)

import numpy as np

x = [1,2,3,4,5]
y = [2,3,789,0,1]

z = np.add(x,y)

print(z)

# No loop is required

# NumPy automatically adds every corresponding element
```

## Create Your Own ufunc

```python
# Normally NumPy provides many ufuncs

# You can also create your own
```

```python
## Step 1: Create a normal function

def myadd(x,y):
    return x+y

## Step 2: Convert it into ufunc

# Syntax -> np.frompyfunc(function, inputs, outputs)
## function	-> Function name
## inputs ->  Number of input arguments
## outputs -> Number of output values

import numpy as np

myadd = np.frompyfunc(myadd,2,1)
```

```python
## Example

import numpy as np

def myadd(x,y):
    return x+y

myadd = np.frompyfunc(myadd,2,1)

print(myadd([1,2,3],[12,3,4]))
```

#### Check if a Function is a ufunc

```python
## Check if a Function is a ufunc

import numpy as np

print(type(np.add))

# <class 'numpy.ufunc'> -> It means add() is a ufunc
```

```python
## Example

print(type(np.concatenate))

# concatenate() is not a ufunc
```

### Using if Statement

```python
## Using if Statement

import numpy as np

if type(np.add) == np.ufunc:
    print("add is ufunc")
else:
    print("add is not ufunc")

```

## Simple Arithmetic ufuncs

```python
## Simple Arithmetic ufuncs

## These functions perform arithmetic on two arrays element by element
```

### 1. add()

```python
#  Adds corresponding elements

import numpy as np

arr1 = np.array([1,2,3,45,5])
arr2 = np.array([1,2,32,78,9])

print(np.add(arr1,arr2))
```

### 2. subtract()

```python
## Subtracts second array from first array

import numpy as np

arr1 = np.array([1,2,3,45,5])
arr2 = np.array([1,2,32,78,9])

print(np.subtract(arr1,arr2))
```

### 3. multiply()

```python
## Multiplies corresponding elements

import numpy as np

arr1 = np.array([1,2,3,45,5])
arr2 = np.array([1,2,32,78,9])

print(np.multiply(arr1,arr2))
```

### 4. divide()

```python
## Divides first array by second array

import numpy as np

arr1 = np.array([1,2,3,45,5])
arr2 = np.array([1,2,32,78,9])

print(np.divide(arr1,arr2))
```

### 5. power()

```python
## Raises first array values to the power of second array values

import numpy as np

arr1 = np.array([1,2,3,5,5])
arr2 = np.array([1,2,2,8,9])

print(np.power(arr1,arr2))
```

### 6. mod() / remainder()

```python
## Returns the remainder after division

import numpy as np

arr1 = np.array([1,2,32,78,9])
arr2 = np.array([1,2,3,45,5])

print(np.mod(arr1,arr2))

print(np.remainder(arr1,arr2))

## np.mod() and np.remainder() produce the same result
```

### 7. divmod()

```python
## Returns both quotient and remainder

import numpy as np

arr1 = np.array([10,20,30])
arr2 = np.array([3,7,9])

print(np.divmod(arr1,arr2))

#In Output:
# First array -> Quotient
# Second Array -> Remainder
```

### 8. absolute()

```python
## Converts negative numbers into positive numbers

arr = np.array([1,-2,33,-56,-9,-32])

print(np.absolute(arr))

# Absolute value simply removes the negative sign
```

## NumPy Rounding Functions

```python
## Rounding Functions

# NumPy provides several ways to round decimal numbers
```

### trunc()

```python
## trunc()

# Removes the decimal part

arr = np.array([23.34, 45.91,-23.0])

print(np.trunc(arr))

# It simply cuts off the decimal part
```

### fix()

```python
## fix()

# Works exactly like trunc()

arr = np.array([23.34, -45.91,23.0])

print(np.fix(arr))
```

### around()

```python
# around()

## Rounds to a specified number of decimal places

# (number, round decimal place number)
print(np.around(3.1116,2))
print(np.around(3.1116,3))
print(np.around(3.1386,2))
```

### floor()

```python
## floor()

# Rounds down to the nearest integer

print(np.floor([3.1, -3.6, 4.9, -5.1]))

# Floor always moves downward
```

### ceil()

```python
## ceil()

# Rounds up to the nearest integer

print(np.ceil([3.1, -3.6, 4.9, -5.1]))

# Ceil always moves upward
```

## ufunc NumPy Logs

```python
# NumPy Logs

# A logarithm answers the question:

# "To what power should the base be raised to get the given number?

## Example
# 2³ = 8
# So
# log₂(8)=3
```

### log2()

```python
## log2()

# Calculates logarithm with base 2

import numpy as np 
arr = np.arange(1,10)
print(np.log2(arr))

# log₂(1)=0

# log₂(2)=1

# log₂(4)=2

# log₂(8)=3
```

### log10()

```python
## log10()

# Calculates logarithm with base 10

arr = np.arange(1,10)

print(np.log10(arr))


# log₁₀(1)=0

# log₁₀(10)=1

# log₁₀(100)=2
```

### log()

```python
## log()

# Calculates the natural logarithm (base e)

arr=np.arange(1,10)
print(np.log(arr))

## e = 2.71828

# This function is widely used in mathematics, statistics, and machine learning
```

### Log with Any Base

```python
# NumPy has no direct function for arbitrary bases

# You can create one using frompyfunc()

from math import log
import numpy as np

nplog = np.frompyfunc(log,2,1)
print(nplog(100,10))

# 10² =100
# Therefore
# log₁₀(100)=2
```
