## Multinomial Distribution

```python
## Multinomial Distribution

# The Multinomial Distribution is an extension of the Binomial Distribution

## Binomial Distribution
# Only 2 possible outcomes

# Example: -> Coin Toss -> Head or Tail

# ---------------------------------------------------------

## Multinomial Distribution
# More than 2 possible outcomes

# Examples: Dice Roll, Blood Groups, Favorite Colors, Survey Responses
# Example (Dice): 1, 2, 3, 4, 5, 6
## Instead of only Head or Tail, we now have 6 possible outcomes
```

### Parameters

```python
# n - Number of experiments - number of times to run the experiment - e.g. roll te dice -> n = 6

# pvals - Probability of each outcome - list of probabilties of outcomes (e.g. [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] for dice roll) -> Each face has equal probability

# size - The shape of the returned array
```

### Example

```python
from numpy import random
x = random.multinomial(n=6,pvals=[1/6]*6)
# x = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6]) # same as
print(x)
```

```python
## Real-life Uses
# -> Dice games
# -> Blood group prediction
# -> Survey categories
# -> Customer choices
# -> Multi-class Machine Learning
```

```python
## Multinomial samples will NOT produce a single value! They will produce one value for each pval
```

## Exponential Distribution

```python
## Exponential Distribution

# It tells us -> How much time until the next event happens

## Exponential distribution is used for describing time till next event e.g. failure/success etc

## Examples:
# -> Time until the next customer arrives
# -> Time until a machine fails
# -> Time until the next phone call
# -> Time until the next earthquake

```

### Parameters

```python
# It has two parameters:

# scale - Average waiting time - inverse of rate ( see lam in poisson distribution ) defaults to 1.0
# Example -> scale = 2 -> means -> Average waiting time ≈ 2 units

# size - The shape of the returned array
```

### Example

```python
x = random.exponential(scale=2,size=(2,3))
print(x)
```

### Visualization

```python
import seaborn as sns

sns.displot(random.exponential(size=1000),kind="kde")

## Many small waiting times, Few large waiting times
```

### Relation Between Poisson and Exponential Distribution

```python
## Poisson distribution deals with number of occurences of an event in a time period whereas exponential distribution deals with the time between these events

## Poisson vs Exponential

## Poisson -> Counts
# How many customers came today?
# Example -> 15 customers

## Exponential -> Waiting Time
# How long until next customer arrives?
# Example -> 3 minutes
```

## Chi-Square Distribution

```python
## Chi-Square Distribution

# Chi Square distribution is used as a basis to verify the hypothesis

# Chi-Square Distribution is mainly used in:
# -> Statistics
# -> Hypothesis Testing
# -> Feature Selection
# -> Machine Learning

# It checks -> Is there a significant relationship?

```

### Parameters

```python
## It has two parameters:

# df - (degree of freedom)
# df = Number of categories - 1 -> Example -> 3 groups -> df = 2

# size - The shape of the returned array
```

### Example

```python
random.chisquare(df=2,size=(2,3))
```

### Visualization

```python
sns.displot(random.chisquare(df=2,size=1000),kind='kde')
```

```python
## Uses
# -> Chi-Square Test
# -> Hypothesis Testing
# -> Feature Selection
# -> Statistics
```

## Pareto Distribution

```python
## Pareto Distribution

# A distribution following Pareto's law i.e. 80-20 distribution (20% factors cause 80% outcome)

## Examples -> Business
# 20% customers -> generate -> 80% sales
```

### Parameters

```python
## It has two parameter:

# a - shape parameter -> Controls the distribution shape

# size - The shape of the returned array
```

### Example

```python
random.pareto(a=2,size=(2,3))
```

### Visualization

```python
sns.displot(random.pareto(a=2,size=1000))

## Many small values, Few very large values
```

```python
## Uses
# -> Wealth distribution
# -> Business
# -> Income analysis
# -> Internet traffic
# -> Sales
```

## Zipf Distribution

```python
## Zipf Distribution

# Zipf distributions are used to sample data based on zipf's law

## Zipf's Law: In a collection, the nth common term is 1/n times of the most common term. E.g. the 5th most common word in English occurs nearly 1/5 times as often as the most common word
```

### Parameters

```python
## It has two parameters:

# a - distribution parameter
# size - The shape of the returned array
```

### Example

```python
random.zipf(a=2,size=(2,3))
```

### Visualization

```python
x=random.zipf(a=2,size=1000)
sns.displot(x[x<10])

## Only values less than 10 are plotted because Zipf can produce very large numbers
```

```python
## Uses
# -> Word frequency in languages
# -> Website popularity
# -> Book sales
# -> Social media followers
# -> Search engine rankings
```

```python
## A common question is: "Which distribution should I use?"

# ->  Multinomial → More than two categories (dice, surveys)
# -> Exponential → Waiting time between events
# - >Chi-Square → Test relationships between categorical variables
# -> Rayleigh → Wireless and signal processing
# -> Pareto → 80/20 rule (wealth, sales, business)
# -> Zipf → Ranking and frequency (words, website visits, popularity)
```
