## Matplotlib Adding Grid Lines

- A grid is a collection of horizontal and vertical lines that makes graphs easier to read.

- Use the grid() function to add grid lines to the plot.

### Example

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.array([80,90,95,100,120])
y = np.array([240,250,290,341,312])

plt.plot(x,y)
plt.grid()
plt.show()

# The graph will display grid lines on both x-axis and y-axis
```

### Display Grid Only on X-axis

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.array([80,90,95,100,120])
y = np.array([240,250,290,341,312])

plt.plot(x,y)
plt.grid(axis='x')
plt.show()
# Only vertical grid lines appear
```

### Display Grid Only on Y-axis

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.array([80,90,95,100,120])
y = np.array([240,250,290,341,312])

plt.plot(x,y)
plt.grid(axis='y')
plt.show()

# Only horizontal lines appear
```

### Change Grid Style

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.array([80,90,95,100,120])
y = np.array([240,250,290,341,312])

plt.title("Example")
plt.plot(x,y)
plt.grid(color = 'green', linestyle='--', linewidth='0.5')
plt.show()
```

## Matplotlib Subplot

- With the subplot() function -> can draw multiple plots in one figure.

- want multiple graphs in one window

- Syntax -> plt.subplot(rows, columns, index)

```python
plt.subplot(1,2,1) #plot1
# the figure has 1 row, 2 columns, and this plot is the first plot

plt.subplot(1,2,2) #plot2
# the figure has 1 row, 2 columns, and this plot is the second plot
```

```python
x = np.array([1,2,4,1,7,8])
y = np.array([12,56,43,21,12,0])
plt.subplot(1,2,1)
plt.plot(x,y)

a = np.array([23,67,54,32,1])
b = np.array([12,34,1,15,6])
plt.subplot(1,2,2)
plt.plot(a,b)
```

## Matplotlib Scatter

- Scatter plot shows individual observations.

- Each point represents one observation.



With Pyplot, -> Use the scatter() function to draw a scatter plot.

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis.

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.array([12,3,56,7,90,1,3,5,17,2])
y = np.array([1,3,5,7,2,12,34,5,8,90])

plt.scatter(x,y)
plt.show()
```

### Compare Plots

### Colors

- Set your own color for each scatter plot with the color or the c argument.

### Color Each Dot

- Set a specific color for each dot by using an array of colors as value for the c argument.

- Note: Cannot use the color argument for this, only the c argument.

```python
x = np.array([5,7,8,7,2,17,2,9,4,13,12,9,6])
y = np.array([99,86,19,108,123,86,100,82,94,78,77,85,86])

colors = np.array(["blue","green","red","yellow","aqua","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x, y, c = colors)

plt.show()
```

### ColorMap

- The Matplotlib module has a number of available colormaps.

- A colormap is like a list of colors, where each color has a value that ranges from 0 to 100

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.show()
```

- This colormap is called `viridis` and as you can see it ranges from 0, which is a purple color, up to 100, which is a yellow color.

####  Include the colormap in the drawing by including the plt.colorbar() statement.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

### Marker Size

- Change the size of the dots with the s argument.

- Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,42,117,2,9,4,111,12,9,6])
y = np.array([99,86,87,48,11,86,13,87,34,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x,y,s=sizes)
plt.show()
```

### Alpha

- Adjust the transparency of the dots with the alpha argument.

- Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,42,117,2,9,4,111,12,9,6])
y = np.array([99,86,87,48,11,86,13,87,34,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x,y,s=sizes,alpha=0.5)
plt.show()
```

### Combine Color Size and Alpha

- Combine a colormap with different sizes of the dots. This is best visualized if the dots are transparent:

```python
## Create random arrays with 100 values for x-points, y-points, colors and sizes:

import numpy as np 
import matplotlib.pyplot as plt 

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))

colors = np.random.randint(100, size=(100))

sizes = np.random.randint(1000, size=(100))

plt.scatter(x,y, c = colors, s = sizes, alpha = 0.5, cmap = 'Dark2')

plt.colorbar()
plt.show()
```

## Matplotlib Bars

```python
### Creating Bars
# With Pyplot, Use the bar() function to draw bar graphs.

import matplotlib.pyplot as plt
import numpy as np 

x = np.array(["A", "B", "D", "A1"])
y = np.array([18,5,34,1])

plt.bar(x,y)
plt.show()
```

- The bar() function takes arguments that describes the layout of the bars.

- The categories and their values represented by the first and second argument as arrays

```python
x = ["GTA", "Delta-Force"]
y = [400, 250]
plt.bar(x,y)
```

### Horizontal Bars

- bars to be displayed horizontally instead of vertically -> use the barh() function

```python
import matplotlib.pyplot as plt
import numpy as np 

x = np.array(["A", "B", "D", "A1"])
y = np.array([18,5,34,11])

plt.barh(x,y)
plt.show()
```

## Matplotlib Histograms

### Histogram

- A histogram is a graph showing frequency distributions.

- It is a graph showing the number of observations within each given interval.

### Create Histogram

- In Matplotlib, -> Use the hist() function to create histograms.

- The hist() function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

- For simplicity -> Use NumPy to randomly generate an array with 250 values, where the values will concentrate around 170, and the standard deviation is 10.

```python
## A Normal Data Distribution by NumPy:

import numpy as np 
x = np.random.normal(170,10,250)
print(x)
```

- The hist() function will read the array and produce a histogram.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170,10,250)

plt.hist(x)
plt.show()
```

## Matplotlib Pie Charts

### Creating Pie Charts

- With Pyplot, Use the pie() function to draw pie charts.

```python
# Pie chart displays percentages

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,23,1,3])

plt.pie(y)
plt.show()

# pie chart draws one piece (called a wedge) for each value in the array
```

- pie chart draws one piece (called a wedge) for each value in the array.
- By default the plotting of the first wedge starts from the x-axis and moves counterclockwise.

- Note: The size of each wedge is determined by comparing the value with all the other values, by using this formula:

- The value divided by the sum of all values: x/sum(x)

### Labels

- Add labels to the pie chart with the labels parameter.

- The labels parameter must be an array with one label for each wedge

```python
y = np.array([35,25,25,15])

my_labels = ["Gta", "Fire", "Vice", "Tommy"]

plt.pie(y,labels=my_labels)
plt.show()
```

### Start Angle

- As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

- The startangle parameter is defined with an angle in degrees, default angle is 0:

```python
y = np.array([35,25,25,15])

my_labels = ["Gta", "Fire", "Vice", "Tommy"]

plt.pie(y,labels=my_labels, startangle=90)
plt.show()
```

### Explode

- Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

- The explode parameter, if specified, and not None, must be an array with one value for each wedge.

- Each value represents how far from the center each wedge is displayed:

```python
y = np.array([35,25,25,15])

my_labels = ["Gta", "Fire", "Vice", "Tommy"]
my_explode = [0.2,0,0,0]
plt.pie(y,labels=my_labels, explode=my_explode)
plt.show()
```

### Shadow

- Add a shadow to the pie chart by setting the shadows parameter to True

```python
y = np.array([35,25,25,15])

my_labels = ["Gta", "Fire", "Vice", "Tommy"]
my_explode = [0.2,0,0,0]

plt.pie(y,labels=my_labels, explode=my_explode, shadow=True)
plt.show()
```

### Colors

- Set the color of each wedge with the colors parameter.

- The colors parameter, if specified, must be an array with one value for each wedge

```python
y = np.array([35,25,25,15])

my_labels = ["Gta", "Fire", "Vice", "Tommy"]
my_colors = ["b", "red", "black", "#1ADF21"]

plt.pie(y,labels=my_labels, colors=my_colors)
plt.show()
```

- Use Hexadecimal color values, any of the 140 supported color names, or one of these shortcuts:

'r' - Red 

'g' - Green

'b' - Blue

'c' - Cyan

'm' - Magenta

'y' - Yellow

'k' - Black

'w' - White

### Legend

- To add a list of explanation for each wedge, use the legend() function.

```python
y = np.array([35,25,25,15])

my_labels = ["Gta", "Fire", "Vice", "Tommy"]

plt.pie(y,labels=my_labels)

plt.legend()

plt.show()
```

### Legend With Header

- To add a header to the legend, add the title parameter to the legend function.

```python
y = np.array([35,25,25,15])

my_labels = ["Gta", "Fire", "Vice", "Tommy"]

plt.pie(y,labels=my_labels)

plt.legend(title = "World Game")

plt.show()
```
