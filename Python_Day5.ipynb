# Python Exception Handling

```python
#Exception -> means Runtime Error
#When a program is running and an error occurs, it is called an exception

#print(10/0) # ZeroDivisionError: division by zero
#The program stops here
#To handle this problem, Python provides try and except

#Syntax
#try:
#    risky code

try:
    print(x) #x is not defined, so it will cause an error

#except
#If an error occurs inside try, except will run
#The program will not crash
except:
    print("error")
#The original error is handled by except

##Without try
#print(x) #-> NameError: name 'x' is not defined
#The program will stop

#       try
#        │
#      Error 
#    /       \
#  Yes        No
#   │           │
# except      else
#   │           │
#   └──────┬────┘
#          │
#       finally

#Specific Exception
#Every error has its own exception name

try:
    print(x)

except NameError:
    print("Variable not exist")

#Error 
try:
    print(10/0)

except NameError:
    print("Name Error")

except:
    print("Some other error")

#Multiple Exception 
try:
    num = 10/0

except ZeroDivisionError:
    print("Cannot divide a number by zero")

except NameError:
    print("Variable Missing")

except:
    print("Unknown Error")

###Else 
#Else runs only if no error occurs inside try

try:
    print("hello")
except:
    print("error")
else:
    print("All okay")

#If an error occurs inside try, else will not run
try:
    print(x)
except:
    print("error")
else:
    print("Done")

#finally
#This is the most important block
#Finally always runs
#It runs whether an error occurs or not

try:
    print(10/0)

except:
    print("Error")

finally:
    print("Program END")

#If no error occurs
try:
    print("Hello")

except:
    print("Error")

finally:
    print("Program End")

#Why use finally?

#Mostly used to:
#-> close files
#-> close database connections
#-> close network connections

try:
    f1 = open("test.txt")

finally:
    f1.close()

#Whether an error occurs or not, file.close() will always run

###Nested Try

#f = open("demo.txt", "r") -> "r" means read mode
#f = open("demo.txt", "w") -> "w" means write mode

try:
    f = open("demo.txt", "w") 
    
    try:
        f.write("Hello")
    except:
        print("Write Error")
    finally:
        f.close()
except:
    print("Open Error")

###raise
#Normally Python creates errors automatically
#But sometimes we want to create our own error

#For this, use raise


age = -2
#if age < 0:
#    raise Exception("Age cannot be negative")

#Exception:
#Age cannot be negative

#This is a custom error created by us


x = "Hello"

#if not type(x) is int:
#    raise Exception("Only Integer Allowed")

#TypeError:
#Only Integer Allowed

###Real Life Example

#ATM 
balance = 500
withdraw = 1000

#if withdraw > balance:
#    raise Exception("Insufficient Balance")

#Exception: Insufficient Balance

#Complete Example
try:
    num = int(input("Enter Number: "))

    result = 100 / num

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Enter numbers only")

else:
    print("Answer =", result)

finally:
    print("Program Finished")


#Keyword        Purpose
#try            Code that may cause an error
#except         Handles the error
#else           Runs only if no error occurs
#finally        Always runs (whether an error occurs or not)
#raise          Creates a custom exception

#try      -> Risky code
#except   -> Handle the error
#else     -> Run if no error occurs
#finally  -> Always run
#raise    -> Create your own exception

```

# Python String Formatting

```python
##String Formatting
#Used to display variables or values nicely inside a string

#F-Strings are the most commonly used method in Python (Python 3.6+)

#Use them when you want to print variables inside a string

#Without formatting
name = "Alisha"

print("Hello", name)

#This also works, but F-Strings are the recommended way.

##With F-String

#Syntax -> f"Text {variable}"

name = "Alisha"
print(f"Hello {name}")

#{} -> placeholder
#Whatever is written inside the braces will be displayed.

#Multiple Variables
name = "ali"
age = 22
passed = True

print(f"My name is {name} and I am {age} years old,-> {passed}.")

#You can also write expressions
a = 10
b = 20

print(f"Sum = {a+b}")

#Multiplication
price = 100
print(f"Double Price = {price*2}")

#Decimal Formatting (.2f)
#Show only 2 digits after the decimal point.
price = 59
print(f"{price: .2f}")

#.2f -> means 2 decimal places

#Round off
num1 = 12.35598
print(f"{num1: .2f}")

#Direct Values
#You do not need to create a variable
print(f"{43: .2f}")

#If Else
#You can also use if-else inside the placeholder

price = 40
print(f"{'Expensive' if price>50 else 'Cheap'}")

#You can also use functions
name = "alisha"
print(f"{name.upper()}")

#You can also use your own function
def square(x):
    return x*x
print(f"Answer = {square(25)}")

#Comma Separator 
#Makes large numbers easier to read

price = 50000000
print(f"{price:,}")

#Common Modifiers
#Modifier    Meaning             Example         Output
#.2f         2 decimal places    12.567          12.57
#:,          Comma separator     500000         500,000
#:>          Right align         f"{10:>5}"     spaces + 10
#:<          Left align          f"{10:<5}"     10 + spaces
#:^          Center align        f"{10:^5}"     centered
#%           Percentage          0.75           75.000000%

###format() Method
#This method was commonly used before Python 3.6

price = 45
print("Price is {}".format(price))

#Multiple Values 
name = "Alisha"
age = 22

print("Name: {} Age: {}" .format(name,age))

##Decimal in format()
price = 45.89542

print("{:.2f}".format(price))

##Index Number
name = "Alisha"
age = 22

print("{1} is {0} years old".format(name,age))
#{0} → name                          {0}, {1}
#{1} → age

#Same Variable Multiple Times
name = "Ali"

print("{0} is good. {0} is intelligent." .format(name))

##Named Indexes 
print("My name is {name}".format(name="ali", age = 87))


#F-String vs format()

#F-String(Recommended)
name = "Alisha"

print(f"Hello {name}")

#format()
name = "ali"
print("hello {}".format(name))
```

# Python Virtual Environment (venv)

```python
# Python Virtual Environment (venv)

# A virtual environment is an isolated Python environment.

# Each project has its own Python interpreter and its own libraries.
# This prevents one project's libraries from affecting another project.

# For example:

# Suppose you have two projects.

# Project 1
# Heart Disease Project

# Requires:
# numpy 1.24
# pandas 2.0
# tensorflow 2.12

# Project 2
# Cyber Crime Project

# Requires:
# numpy 2.1
# pandas 2.3
# flask 3.1

# If both versions are installed in the same Python environment,
# version conflicts may occur.

# That is why we use a virtual environment.

# Project Structure

# │
# ├── Project A
# │      numpy 1.24
# │      pandas 2.0
# │
# └── Project B
#        numpy 2.1
#        pandas 2.3

# Without Virtual Environment

# Suppose you install:
%pip install numpy==1.24

# Later, for another project:
%pip install numpy==2.1

# Now the first project will also use NumPy 2.1.
# This may cause compatibility issues or errors.

# With Virtual Environment

# Project A
# numpy 1.24

# Project B
# numpy 2.1

# Both projects work independently without conflicts.

# Command to create a Virtual Environment
# Open Command Prompt or Terminal.

# %python -m venv myproject

# python      -> Runs the Python interpreter.
# -m          -> Runs a Python module.
# venv        -> Virtual environment module.
# myproject   -> Name of the virtual environment folder.

# Result
# A new folder will be created.

# myproject/
#     Include/
#     Lib/
#     Scripts/
#     pyvenv.cfg

# This folder is created automatically.
# What do these folders contain?

# Include      -> Header files (usually ignored).
# Lib          -> Installed packages (NumPy, Pandas, Flask, etc.).
# Scripts      -> Activation files (activate, activate.bat, python.exe, pip.exe).
# pyvenv.cfg   -> Configuration file (Python version and environment settings).

# Activate the Virtual Environment
# Windows

# %myproject\Scripts\activate

# or

# %.\myproject\Scripts\activate

# Output
# (myproject)
# C:\Users\Alisha>

# The text inside the brackets:
# (myproject)

# Indicates that the virtual environment is active.

%pip install cowsay

import cowsay

cowsay.cow("Hello")

# Another Project
# python -m venv AIProject

# Activate
# AIProject\Scripts\activate

# Install
# pip install tensorflow

# TensorFlow will be installed only inside the AIProject environment.

# Deactivate the Virtual Environment
# deactivate

# The virtual environment is now turned off.

# If you run:
# python test.py

# And the file contains:
# import cowsay

# You will get:
# ModuleNotFoundError
# No module named 'cowsay'

# Because cowsay was installed only inside the virtual environment.
# Activate it again

# myproject\Scripts\activate

# Now run:
# python test.py

# It will run successfully.

# Delete a Virtual Environment
# Simply delete the virtual environment folder.

# Or use the command:
# rmdir /s /q myproject

# The entire virtual environment folder will be deleted.

# Why do we only delete the folder?
# Because all installed packages are stored inside that folder.

# Nothing is removed from the system-wide Python installation.

#### Common Commands

# 1. Create
# -> python -m venv myproject

# 2. Activate
# -> myproject\Scripts\activate

# 3. Install a Package
# -> pip install numpy

# 4. Check Installed Packages
# -> pip list

# 5. Uninstall a Package
# -> pip uninstall numpy

# 6. Deactivate
# -> deactivate

# 7. Delete the Virtual Environment
# -> rmdir /s /q myproject
```

# Python  __ init __ () Method

```python
# __init__() 

# __init__() is a special method called a constructor
# It runs automatically when an object is created

# __init__() sets the initial values of an object

###Without __init__():
class Person:
    name = ""
    age = 12
    
p1 = Person()

p1.name = "alisha"
p1.age = 22

print(p1.name)
print(p1.age)

#Here, values are assigned manually after creating the object

#With __init__()
class Person1:
    
    def __init__(self, name, age): #Python automatically calls this function when an object is created
        self.name = name
        self.age = age

p1 = Person1("alisha", 22)
print(p1.name)
print(p1.age)

#Here, values are assigned while creating the object

#What is self? 
#self represents the current object

#Object -> p1 = Person("Alisha",22)

#Python internally: -> Person.__init__(p1,"Alisha",22)

#self = p1
#name = "Alisha"
#age = 22

#So, self always refers to the current object

#self.name = name -> Store the user value inside the object

###Multiple Objects
class multi:
    def __init__(self,name):
        self.name = name

m1 = multi("ali")
m2 = multi("alisha")

print(m1.name, m2.name)

#Default Values
class Default1:
    def __init__(self,name,age=18):
        self.name = name
        self.age = age

d1 = Default1("alis") #Age is not given, so the default value is used
print(d1.name, d1.age)

d2 = Default1("Sajjad",25) #Here the user's value is used
print(d2.name, d2.age) 

#Multiple Parameters
class Multip:
    def __init__(self,name,age,city,country):
        self.name = name
        self.age = age
        self.city = city
        self.country = country

p1 = Multip("Alisha",22,"Sahiwal","Pakistan")
print(p1.name)
print(p1.city)

####EXample
class Student_1:

    def __init__(self,name,roll,dept):
        self.name1=name
        self.roll1=roll
        self.dept1=dept

s1 = Student_1("Alisha",101,"CS")

print(s1.name1)
print(s1.roll1)
print(s1.dept1)

#Important Points
# __init__() is called a constructor
# It is called automatically when an object is created
# It assigns the initial values.
# self represents the current object
# self.name = name stores the value inside the object

###Example Exercise
# Create the Dog class
class Dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def bark(self):
        print(self.name, "says Woof!")
# Create an object
d1 = Dog("Buddy", 3)
# Call the bark method
d1.bark()

```
