# Python Class Properties

```python
# Property = Variable that stores data inside an object 
# For example: name, age, city -> these are properties

class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age 

# self.name and self.age are properties

p1 = Person("Alisha", 34)   #p1 = self #Object
print(p1.name)
print(p1.age)

###Access Properties 
# Use the dot (.) operator to access object properties

class Car:
    def __init__(self,brand,model):
        self.brand = brand
        self.model = model

c1 = Car("Toyota", "Corolla")

#Access
print(c1.brand)
print(c1.model)

##Modify Properties 
# We can also change the value of a property

class Person1:
    def __init__(self,name,age):
        self.name = name
        self.age = age

p1 = Person1("Ali", 23)
print(p1.age) #Output -> 23

#Change Value
p1.age = 21
print(p1.age) #Output -> 21

###Delete Properties 
# We can also delete a property

class Person2:
    def __init__(self,name,age):
        self.name = name
        self.age = age

p1 = Person2("alisha", 33)
print(p1.name, p1.age)

#delete property
del p1.age
#print(p1.name, p1.age) #AttributeError: 'Person2' object has no attribute 'age'

###Class Property vs Object Property 

#1. Object(Instance) Property 

# Every object has its own value
class Person4:

    def __init__(self,name):
        self.name=name

# Each object has a different name
# This is an instance property
p1 = Person4("ali")
p2 = Person4("alisha")

print(p2.name)

#2. Class Property 

# A class property is written inside the class
# but outside all methods

class Person5:
    
    species = "Human" #class Property  
    
    def __init__(self,name):
        self.name = name 

# Both objects share the same class property
p1 = Person5("ali1")
p2 = Person5("alisha1")

#   Person
# -------------
# species = Human
#        |
# -------------------
# |                 |
# p1               p2
# name=Ali         name=alisha

print(p1.species)
print(p2.species, p2.name)


### Change Class Property

class Person6:
    lastName = ""
    
    def __init__(self,name):
        self.name = name

p1 = Person6("Sajjad")
p2 = Person6("alisha")

#ClassName.PropertyName = "value"
Person6.lastName = "ALi"

print(p1.name, p1.lastName)
print(p2.name, p2.lastName)

# Add a New Property

#Suppose 
class Student1:
    def __init__(self,name):
        self.name = name


s1 = Student1("sajjad")
s2 = Student1("ali")

s1.age1 = 33 #New property
s1.city1 = "Lahore" #New property
print(s1.name)
print(s1.age1)
print(s1.city1)

# Will s2 also have age1 and city1? -> #No
s2 = Student1("ali")
#print(s2.age) #AttributeError
# These properties were added only to s1

# Instance Property                      Class Property
# self.name                           species = "Human"
# Written inside __init__()           Written inside the class but outside methods
# Each object has its own value       Shared by all objects
#p1.name != p2.name              	p1.species == p2.species 

# self.variable -> Instance Property -> Each object has its own value 
# variable written inside the class but outside methods -> Class Property -> Shared by all objects

#exercise Example
# Create the Student class
class Student:
    def __init__(self,name,grade):
        self.name = name
        self.grade = grade

# Create an object
s1 = Student("Anna", "A")
# Print the grade
print(s1.grade)
# Change the grade
s1.grade = "B"
# Print the updated grade
print(s1.grade)

```

# Python Class Methods

```python
# Python Class Methods 

# Class Method -> A method is a function defined inside a class 

# Outside a class -> Function 
# Inside a class -> Method

class New_1: #new_1 -> class
    def greet(self): #greet() -> method 
        print("hello")

# How to call a method

n1 = New_1()

n1.greet() # Method call    # Python automatically writes -> New_1.greet(n1)
#n1.greet() = New_1.greet(n1)
# That is why self is required

# Why do we use self?

class Person_new:
    def __init__(self,name):
        self.name=name

    def greet(self):
        print(self.name)

p1=Person_new("Alisha") #object
# Here self means p1 -> self.name = p1.name

print(p1.name)

### Methods can also take parameters just like functions

class Calculator:
    def add(self,a,b):
        return a + b

#object
c = Calculator()
#call
print(c.add(8,9)) #python internally -> Calculator.add(c,5,3)

### Methods can also change object properties

class Change1:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def birthday(self):
        self.age = self.age + 1

b1 = Change1("Ali", 34)
print(b1.age)
b1.birthday()
print(b1.age) # The method updated the age

### str() Method

# This is a special method (magic method)

# print(object) -> Python automatically calls __str__()

# Without __str__()

class Person_1:

    def __init__(self,name):
        self.name=name

p1=Person_1("Ali")

print(p1) # It prints only the object address -> <__main__.Person_1 object at 0x...>

# With __str__()
class New_2:
    def __init__(self,name):
        self.name = name
    
    def __str__(self):
        return self.name

#Object 
p1 = New_2("ali")
print(p1) # Now it prints the actual output

# Python internally changes print(p1) into print(p1.__str__())

#Example
class Data_1:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
    
    def __str__(self):
        return f"{self.name} got {self.marks} marks"

d1 = Data_1("ali", 56)
print(d1)

### Multiple Methods
# We can create as many methods as we want in one class

class Data_2:
    def __init__(self, name):
        self.name = name
    
    def method1(self):
        print("hello")
    
    def method2(self):
        print("welcome", self.name)
    
    def method3(self):
        print("Done Now")

d1 = Data_2("alisha")
d1.method1()
d1.method2()
d1.method3()

# Each method performs a different task

##Playlist Example

class Playlist:
    def __init__(self,name):
        self.name = name
        self.songs=[]   #initially empty list
    
    def add_song(self,song):
        self.songs.append(song)
    
    def show_songs(self):
        return self.songs
    
    def remove_song(self,song):
        if song in self.songs:
            self.songs.remove(song)
            print(f"Removed: {song}")
        

p1 = Playlist("Favorites")
p1.add_song("song_A")
p1.add_song("song_B")
print(p1) #object address print

#show songs
p1.show_songs() #list of songs

#Remove songs
p1.remove_song("song_A")

#update show
p1.show_songs()

###Delete Method 
# A method can also be deleted

class De1:

    def greet(self):
        print("Hello")

#delete method
del De1.greet

#object 
p1 = De1()
#p1.greet() #error -> AttributeError

# Because the greet method no longer exists in the class.


#Exercise 
# Create the Rectangle class
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width*self.height

# Create an object
r1 = Rectangle(5,3)
# Print the area
print(r1.area())
```

# Python Inheritance

```python
# Python Inheritance

# One class can reuse the variables and methods of another class 
# This helps avoid writing the same code again

# Imagine there is a Person

# Every person has:

#First Name
#Last Name
#Print Name method

# A Student is also a Person

# So there is no need to write the same code again

# Simply inherit Student from Person

#Person
#   │
#   │
#Student

# Student automatically gets all the properties and methods of Person

class Person_new1:
    def __init__(self,fname,lname):
        self.fname = fname
        self.lname = lname
    
    def printname(self):
        print(self.fname, self.lname)

#Parent Class = Person_new1
#Variables -> firstname , lastname
#method -> printname()

obj1 = Person_new1("ali", "alisha")
obj1.printname()

#Child Class
# In the class syntax, pass the parent class as a parameter
class Student_1(Person_new1): # Student_1 inherits from Person_new1, so it automatically gets fname, lname, and printname()
    pass
# pass -> Nothing is written in this class yet
# It only uses everything from the parent class

s1 = Student_1("sajjad", "ali")
s1.printname()

# We did not write anything inside Student, but it still works because of inheritance

## Add __init__() in Child

# Suppose Student has its own extra information 
# For example: year

class Student_2(Person_new1):
    def __init__(self, fname, lname):
        pass

# Problem -> Python ignores the parent's __init__() 
# This is called overriding

## Override

#Parent 
class Parent:

    def __init__(self):
        print("Parent")

#Child 
class Child(Parent):

    def __init__(self):
        print("Child")

Child() #Output : Child 
# The parent's constructor was not executed
# The child replaced it

## Call the Parent's __init__()

class Parent1:

    def __init__(self,name):
        print("Parent1")

#Child 
class Child1(Parent1):

    def __init__(self,name):
        Parent1.__init__(self,name) # This line calls the parent's constructor

c2 = Child1("ali")

# First Student.__init__() runs
# Then Person.__init__() runs 
# Finally, name = "Ali" is stored

####super() 

# super() does the same job in an easier way

#Instead of -> Person.__init__(self,name)

#use  ->  super().__init__(name)

# Both do the same thing
# Only the syntax is different

class Parent2:
    def __init__(self,name):
        self.name = name

class Child2(Parent2):
    def __init__(self,name):
        super().__init__(name)

c = Child2("alisha")
print(c.name)

###New Add a New Property

# Suppose the child class has extra information

class Child3(Parent2):
    def __init__(self,name, year):
        super().__init__(name)
        self.year = year

c3 = Child3("alisha", 2020)
print(c3.name, c3.year)

# The object now has:
#name = alisha
#year = 2020

###NAdd a New Method
class Child4(Parent2):
    def __init__(self,name, year):
        super().__init__(name)
        self.year = year
    
    def welcome(self):
        print("welcome", self.name)

c4 = Child4("alisha", 2020)
c4.welcome()

###Parent Method Override
class Parent4:
    def __init__(self,name):
        self.name = name
    def printname(self):
        print(self.name)

class Child5(Parent4):
    def printname(self):
        print("Child:",self.name)

s = Child5("ali")
s.printname()

# The parent method was not executed
# This is called Method Overriding


###Example
class last1:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def info(self):
        print(self.name,self.age)

class last_2(last1):
    def __init__(self,name,age,roll_no):
        super().__init__(name,age)
        self.roll_no = roll_no
    
    def display(self):
        print(self.name)
        print(self.age)
        print(self.roll_no)

s1 = last_2("alishaasajjad", 23, 373)

s1.info()
s1.display()

#Exercise Example
# Create the Animal class
class Animal:
    def __init__(self,name):
        self.name = name
    def speak(self):
        print(self.name) 

# Create the Dog class (inherits from Animal)
class Dog(Animal):
    pass

# Create an object
d1 = Dog("Rex")
# Call the speak method
d1.speak()
```

# Python Polymorphism

```python
# Python Polymorphism

#poly = Many
#Morph = Forms

# The same method/function name can work differently for different objects

# Method -> move() 

# # Car -> Drive 
# # Boat -> Sail 
# # Plane -> Fly 

#The method name is the same, but the output is different
#This is called polymorphism

#1. Function Polymorphism
# Python built-in functions can also be polymorphic

x = "Hello"
print(len(x)) # 5 -> String character count

#Tuple
t = (10,20,30,40)
print(len(t)) # 4 -> Tuple items count

#Dictionary 
d = {
    "name" : "alisha",
    "age" : 23
}
print(len(d)) # 2 ->Keys count

#Notice -> len() is the same everywhere
#String -> characters
#Tuple -> items 
#Dictionary ->key-value pairs 

# This is called Function Polymorphism


#2. Class Polymorphism

class Car:

    def move(self):
        print("Drive")

class Boat:
    def move(self):
        print("Sail")

class Plane:
    def move(self):
        print("fly")

##Objects
car = Car()
boat = Boat()
plane = Plane()

#Loop
for x in (car,boat,plane):
    x.move()

#Output 
#Drive
#Sail
#Fly

# We called move() for every object 
# Python automatically decides which move() method to execute 
# This is called Polymorphism



###Inheritance Polymorphism

#Parent
class Vehicle:
    def move(self):
        print("Move")

#Child
class Car1(Vehicle): # Car does not have its own move() method, so it uses the parent's method
    pass

class Boat1(Vehicle): # Boat has its own move() method, so it uses its own method
    def move(self):
        print("Sail")

class Plane1(Vehicle):
    def move(self):
        print("Fly")

#Objects
car =Car1()
boat = Boat1()
plane = Plane1()

for x in (car, boat, plane):
    x.move()

#Output
#Move
#Sail
#Fly


## How did overriding happen?

#Parent
class Vehicle1:
    def move(self):
        print("Move")

#Child
class Boat2(Vehicle1):
    def move(self):
        print("Sail")

# Boat created a method with the same name 
# So the parent's method was replaced 
# This is called Method Overriding 
# It is also a part of Polymorphism

###Example
class Student11:
    def intro(self):
        print("I am a Student")

class Teacher:
    def intro(self):
        print("I am a Teacher")

class Doctor:
    def intro(self):
        print("I am a Doctor")

people = [Student11(),Teacher(),Doctor()]

for p in people:
    p.intro()


###Exercise Example
# Create the Cat class
class Cat:
    def sound(self):
        print("Meow")

# Create the Fox class
class Fox:
    def sound(self):
        print("Wa-pa-pa-pa-pa!")

# Create objects and loop
c1 = Cat()
f1 = Fox()

for x in (c1, f1):
    x.sound()

```

# Python Encapsulation

```python
#Python Encapsulation

# Encapsulation means keeping data (variables) and methods inside one class 
# and protecting the data from direct access 

# # ATM -> We can only withdraw or deposit money 
# We cannot directly access the ATM's internal system 

# This helps keep the data secure

###Example without Encapsulation

class Encap1:
    def __init__(self):
        self.age = 23

e = Encap1()
print(e.age)

e.age = -100
print(e.age)

# Problem -> The user changed the age to -100
# This is an invalid value, so we use encapsulation

#Private Properties

# In Python, use double underscores (__) to create a private variable

class Encap2:
    def __init__(self,name,age):
        self.name = name #public
        self.__age = age #private

e1 = Encap2("ali", 32)
print(e1.name)
# print(e1.__age) #error -> AttributeError
# Because age is private, it cannot be accessed directly

###Getter Method
# A getter method is used to read a private value

class Encap3:

    def __init__(self,name,age):
        self.name = name
        self.__age = age

    def get_age(self):
        return self.__age

p = Encap3("Alisha",22)
print(p.get_age()) # Now the age is displayed

###Setter Method
# A setter method is used to change the value of a private variable

class Encap4:
    def __init__(self,name,age):
        self.name = name
        self.__age = age
    
    def set_age(self,age):
        self.age = age
    
    def get_age(self):
        return self.__age

a = Encap4("ali",45)

a.set_age(15)
print(a.get_age())

##Validation 
## The biggest advantage of a setter is that we can validate the data

class Encap5:
    def __init__(self,name,age):
        self.name = name
        self.__age = age
    
    def set_age(self,age):
        if age > 0 :
            self.__age = age
        
        else:
            print("Invalid age")
    
    def get_age(self):
        return self.__age

b = Encap5("ali",35)

b.set_age(-98)
print(b.get_age()) ## The age was not changed because the condition is false
# This is why encapsulation is useful

#Example
class Studentt:

    def __init__(self,name):
        self.name = name
        self.__grade = 0
    
    def set_grade(self,grade):
        if 0<=grade<=100:
            self.__grade = grade
        else:
            print("Invalid Grade")

    def get_grade(self):
        return self.__grade
    
    def get_status(self):
        if self.__grade >= 60:
            return "Passed"
        else:
            return "Failed"

s = Studentt("Alisha")

s.set_grade(75)

print(s.get_grade())

print(s.get_status())

print(s.set_grade(908))

###Protected Properties 
#(_) Single Underscore

class Protect1:
    def __init__(self,name,salary):
        self.name = name 
        self._salary = salary

p = Protect1("alisha", 8948)
print(p._salary)
# This will work
# But the underscore means: 
# "Please do not use this variable directly from outside the class" 

# Python does not block it
# It is only a warning or naming convention

##Private MEthod
# Methods can also be private

class Calculator:
    def __validate(self):
        print("Calculating")

c = Calculator()
#c.__validate() #AttributeError

## A private method can only be called inside the class

class Calculator1:
    def __validate(self,num):
        return isinstance(num,(int,float))
    
    def add(self,num):
        if self.__validate(num):
            print("valid")
        else:
            print("Invalid")

c = Calculator1()
c.add(5)   # add() calls the __validate() method
        

###Name Mangling
# Python actually renames private variables internally

#__age -> Python Internally -> _Person__age

class Name1:
    def __init__(self):
        self.__age = 22

n = Name1()
#print(n.__age)  #AttributeError
print(n._Name1__age) # This works
# This is called Name Mangling

# This is only Python's internal mechanism
# In real projects, you should not use this,
# because it defeats the purpose of encapsulation

#Exercise Example

# Create the ScoreBoard class
class ScoreBoard:
    def __init__(self,score):
        self.__score = score

    def get_score(self):
        return self.__score 

# Create an object
s1 = ScoreBoard(0)
# Print the score
print(s1.get_score())

```

# Python Inner Classes

```python
# Python Inner Classes

# Inner Class -> A class created inside another class 

#Syntax 
class Outer: #Outer Class
    
    class Inner:  #Inner Class
        pass

#Outer Class
#│
#├── Variables
#├── Methods
#│
#└── Inner Class
#      │
#      ├── Variables
#      └── Methods

#Example 1
class Outer1:
    def __init__(self):
        self.name = "Outer CLass"
    
    class Inner:
        def __init__(self):
            self.name = "Inner Class"
        
        def display(self):
            print("This is Inner Class")

#Objects
outer = Outer1()
print(outer.name)
# Here, only the outer class object is created
# The inner class object is not created yet

#Inner class object

# Before using the inner class,
# we must first create an outer class object

outer = Outer1()
inner = outer.Inner()
print(inner.name)
inner.display()

#Outer Object
#      |
#      |
#   Inner()
#      |
#display()

## Inner Class cannot access outer variables automatically

# The inner class does not automatically get the outer class variables

class Outer2:
    def __init__(self):
        self.name = "alisha"
    
    class Inner:
        def show(self):
            print(self.name)

outer = Outer()
inner = outer.Inner()
#inner.show()  #AttributeError ## Because self refers to the inner class object, not the outer class object

##Outer Object Pass
class Outer3:
    def __init__(self):
        self.name = "ali"
    
    class Inner:
        def __init__(self,outer):
            self.outer = outer
        def show(self):
            print(self.outer.name)

outer = Outer3()
inner = outer.Inner(outer)
inner.show()

#Outer Object
#      |
#      |
#Pass to Inner
#      |
#self.outer
#      |
#outer.name

##Practical Example
class Car1:
    def __init__(self,brand,model):
        self.brand = brand
        self.model = model
        
        self.engine = self.Engine()
    
    class Engine:
        def __init__(self):
            self.status = "off" #initially engine off ha
        
        def start(self):
            self.status = "Running"
            print("Engine Started")
    
        def stop(self):
            self.status = "Off"
            print("Engine Stopped")
    
    def drive(self):
        if self.engine.status=="Running":
            print("Driving...")
        else:
            print("Start Engine First")


car1 = Car1("Toyota", "Corolla")
car1.drive()
car1.engine.start()
car1.drive()

##Multiple Inner CLasses
# One outer class can have multiple inner classes

class Computer:
    def __init__(self):
        
        self.cpu = self.CPU()
        self.ram = self.RAM()
        
    class CPU:
        def process(self):
            print("Processing")
    
    class RAM:
        def store(self):
            print("Storing")

#Object
c = Computer()
c.cpu.process()
c.ram.store()

# When do we use an Inner Class?

# When one class is used only in the context of another class

#Examples:

#Car → Engine
#Computer → CPU
#University → Department
#House → Room

```
