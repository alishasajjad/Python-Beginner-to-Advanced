# PYTHON JSON

```python
###Python JSON
#JSON -> JavaScript Object Notation

#Python, java, C#, PHP, NodeJS -> all use JSON

#JSON -> data store and send data from one computer to another computer

#Real Life Example 
#Suppose Facebook server sent a profile
#{
#    "name":"Alisha",
#    "age":22,
#    "city":"Sahiwal"
#}
#      ^
# JSON |
#python understand this

#PYTHON and JSON Dictionary are same

#Python Dictionary
student = {
    "name" : "Alisha",
    "age" : 123,
    "Passed" : True,
    "marks" : [12,34,56]
}

#JSON Dictionary 
{
    "name" : "Alisha",
    "age" : 123,
    "Passed" : True,
    "marks" : [12,34,56]
}

#Dictionary = Python Object
#JSON = String (Text Format)

#Python Dictionary	   JSON
#Python object	    Text/String
#Python use      	for Data transfer 

#JSON Module 
#in python -> when use json -> first import module
import json 

#JSON has two important functions
#loads(), dumps() 

#1. json.loads()
#json string -> Python Object 

#json covert the string into python dictionary 

#loads -> load string 

import json 
x = '{"name": "Alisha", "age":20}' #string
y = json.loads(x) #convert into dictionary
print(y)

##Now access items as dictionary
print(y["name"])

#2.  json.dumps() 
#opposite of loads
#Python Object -> JSON STring
#dump : dictionary -> JSON

import json
student = {
    "name" : "Alisha",
    "age" : 123,
    "Passed" : True,
    "marks" : [12,34,56]
}
z = json.dumps(student)
print(z)

###JSON Support Which Data Types
# many objects in python which convert into jSON

#Dictionary 
import json
print(json.dumps({"name":"Ali", "age": 12}))

#List 
print(json.dumps([1,2,3,4,"ali"])) # list in json -> array

#Tuple
print(json.dumps((1,2,3,4,"ali"))) #tuple in json -> array

#String
print(json.dumps("Hello"))

#Integer
print(json.dumps(23))

#Float
print(json.dumps(12.98))

#Boolean 
print(json.dumps(True))

#None 
print(json.dumps(None))

##Python	JSON
#dict	   Object
#list	   Array
#tuple	   Array
#str	   String
#int	   Number
#float	   Number
#True	   true
#False	   false
#None	   null

#JSON(Problem) -> not output readable
#data all in one line -> so use indent to make output data beautiful

#indent
import json 

student = {
    "name" : "Alisha",
    "age" : 123,
    "Passed" : True,
    "marks" : [12,34,56]
}

print(json.dumps(student,indent = 3))
print(json.dumps(student,indent = 8))

###sort_keys = True 
#keys alphabetically arrange -> of dictionary 

student = {
    "name" : "Alisha",
    "age" : 123,
    "Passed" : True,
    "marks" : [12,34,56]
}
print(json.dumps(student,indent = 4,sort_keys=True))

###Separators 
#Normally JSON -> ""name" : "ali"

#use separator #formatting change -> less practical use
print (json.dumps(student, separators=(".", "=")))



###Real Life example
#Data From server
#JSON
{
    "name":"Alisha",
    "cgpa":3.1
}

#Python
import json

data='{"name":"Alisha","cgpa":3.1}'

student=json.loads(data)

print(student["name"])
print(student["cgpa"])

#OUTPUT 
#ALisha
#3.1


###Suppose now we send data to server
#Python
student={
    "name":"Ali",
    "cgpa":3.8
}

#JSON (Server)
data=json.dumps(student)
print(data)
#now this is possible to send data to server


```

# PYTHON REGEX (REGULAR EXPRESSION)

```python
###PYTHON RegEx ->Regular Expression
#RegEx -> powerful tool of python 
#which is used to search, check, replace,split and validate in a string(text)

#RegEx -> Smart Pattern to find a text
#Email check 
#Phone number validate 
#Password check 
#word find in Text
#Spaces remove 

txt = "I am a Student from Comsat Sahiwal"
# check "word" is in text or not
# check Sahiwal is in end
# check how many times "a" in txt
# remove Spaces

#Module Import #before using RegEx, first import module
import re

#RegEx has mostly 4 important function
#1. findall()
#2. search()
#3. split()
#4. sub()

#1. findall()

#Syntax -> re.findall(pattern, string)
import re

txt = "I am a Student from Comsat Sahiwal"
find1 = re.findall("Sahi", txt)
print(find1)

#if nothing found
txt = "I am a Student from Comsat Sahiwal"
find2 = re.findall("Wahi", txt) #Empty list return 
print(find2) 

#2. search()
#it search only first match and return object
txt = "I am a Student from Comsat Sahiwal, BSCS, FA23-BCS-373"
find3 = re.search("CS", txt)
print(find3)

#Position find -> first occurrence
txt = "I am a Student from Comsat Sahiwal, BSCS, FA23-BCS-373"
find4 = re.search("CS", txt)
print(find4.start()) 

#if match not found -> output -> None
txt = "I am a Student from Comsat Sahiwal, BSCS, FA23-BCS-373"
find5 = re.search("CSC", txt)
print(find5) 

#3.split()
#it break string

#syntax -> re.split(r"pattern", string)
import re
txt = "I am a Student from Comsat Sahiwal, BSCS, FA23-BCS-373"
x = re.split(r"\s", txt) #always use raw string r"" when write a RegEx
print(x)

#Split only on first space 
import re
txt = "I am a Student from Comsat Sahiwal, BSCS, FA23-BCS-373"
x = re.split(r"\s", txt, maxsplit=1)  
print(x)

#4. sub()
#it replace

#Syntax -> re.sub(pattern, replacement, string)
import re
txt = "I am a Student from Comsat Sahiwal, BSCS, FA23-BCS-373"
x = re.sub(r"\s", "-", txt)  
print(x)

#only first 2 spaces replace
import re
txt = "I am a Student from Comsat Sahiwal, BSCS, FA23-BCS-373"
x = re.sub(r"\s", "-", txt, count=2) #use count   
print(x)

###META CHARACTERS
#special symbols of RegEx

#1. ^ 
#means -> starts with ->a txt start with
import re

txt = "Hello World"
print(re.search("^Hell", txt))  #Match object

print(re.search("^World", txt)) #None ->Because word is not in start

#2. $
#means ends with
import re

txt = "Alisha Sajjad"
print(re.search("saj$", txt))  #None 

print(re.search("jad$", txt)) #match object

#3. . 
#means -> any one character
print(re.findall("ali.sha", "aliasha aliisha aliisha aliushat"))
#output -> ['aliasha', 'aliisha', 'aliisha', 'aliusha']

print(re.findall("ali.sha", "as sa sl al")) #return empty List
print(re.findall("ali.sha", "alia isha")) #return empty

#4. *
# means 0 or more
#pattern ab*
#means -> a, ab, abbb, abbbb etc

#5. +
#atleast 1 or more
#pattern ab+
#means -> ab, abb, abbb, abbb ect

#6. ?
#means 0 or 1
#colou?r
#color, colour -> both accept

#7. []
#character se
#means [abc]
#a or b or c

print(re.findall("[as]","alisha"))

###Range 
#[a-z] ->small letters
#[A-Z] ->capital letters
#[0-9] ->Digits

import re

# Important Special Sequences

# 1. \d
#Means -> Digit
#Matches -> 0-9
print(re.findall(r"\d", "abc123"))
# Output: ['1', '2', '3']


# 2. \D
#Means -> Not Digit
print(re.findall(r"\D", "abc123"))
# Output: ['a', 'b', 'c']


# 3. \s
#Means -> Whitespace
#Matches:
#Space
#Tab
#Newline
print(re.findall(r"\s", "Hello World"))
# Output: [' ']


# 4. \S
#Means -> Not Whitespace
print(re.findall(r"\S", "Hello World"))


# 5. \w
#Means -> Word Characters
#Matches:
#A-Z
#a-z
#0-9
#_
print(re.findall(r"\w", "Hello_123!"))


# 6. \W
#Means -> Not Word Characters
print(re.findall(r"\W", "Hello_123!"))
#Output: ['!']


# 7. \b
#Means -> Word Boundary
print(re.search(r"\bcat", "cat is here"))
#Match
print(re.search(r"\bcat", "mycat is here"))
#No Match


# 8. \A
#Means -> Beginning of String
print(re.search(r"\AHello", "Hello World"))
#Match


# 9. \Z
#Means -> End of String
print(re.search(r"World\Z", "Hello World"))
#Match


### Flags
# They change search behavior

# 10. re.I
# Ignore Case

txt = "ALISHA"
print(re.search(r"alisha", txt, re.I))
# Match

# Without re.I
print(re.search(r"alisha", txt))
# No Match


# 11. re.M
# Multi-line Mode
# ^ and $ work for each line

txt = "Hello\nWorld"

print(re.findall(r"^World", txt, re.M))
# Output: ['World']


# 12. re.S
# Dot (.) also matches newline
# Normally '.' does not match newline

txt = "Hello\nWorld"

print(re.search(r"Hello.World", txt, re.S))
# Match

#### Match Object


# When re.search() finds a match,
# it returns a Match Object.
txt = "I am a Student from Comsat Sahiwal, BSCS, FA23-BCS-373"
x = re.search(r"am", txt)
print(x)
# Output: <re.Match object ...>

print(x.group())
# Output: am

print(x.start())
# Output: 2

print(x.end())
# Output: 4

print(x.span())
# Output: (2, 4)


###MATCH OBJECT
#when we find a object after search -> called match object
import re
txt = "I am a Student from Comsat Sahiwal, BSCS, FA23-BCS-373"
x = re.search( "am", txt)  
print(x)

#output -> <re.Match object>

#group -> return matched text
print(x.group())

#start()
#start index 
print(x.start())

#end()
#end index
print(x.end())

#span()
#start and end both index
print(x.span())

#string
#original string return
print(x.string)


#PHONE NUMBER DIGITS
import re
txt = "Phone: 092309983290"
print(re.findall(r"\d", txt))

##Email CHeck 
import re
email = "äbc@gmail.com"
print(re.search("@gmail.com", email))

##Spaces Replace
import re
txt = "ahmad ibrahim"
print(re.sub(r"\s", "_",txt))

##Sentence Split
import re
txt = "Abc Def GH"
print(re.split(r"\s", txt, maxsplit=1))

###Function           Purpose
#findall()	    return all matches (list)
#search()	    search only first match 
#split()	    break string on pattern
#sub()	        replace the pattern

##Symbol	Meaning
# ^	       Start of string
# $	       End of string
# .	       any one character
# *	       0 or more occurrences
# +	       1 or more occurrences
# ?	       0 or 1 occurrence
# []	   Character set
# \d	   Digit (0–9)
# \D	   Not digit
# \s	   Whitespace (space, tab, newline)
# \S	   Not whitespace
# \w	   Word character (A-Z, a-z, 0-9, _)
# \W	   Not word character
# \b	   Word boundary

import re

#Check if the string starts with "The" and ends with "Spain":

txt = "The rain in Spain"
x = re.search("^The.*Spain$", txt)

if x:
    print("YES! We have a match!")
else:
    print("No match")


```
