# String Methods

```python
###String Methods
# All String methods return new values. They do not change the original string

#1. Case Conversion Methods
#upper(), lower()
#Capitalize() -> first character capital
txt = "alisha sajjad"
print(txt.capitalize())

#title()-> first letter of every word in string 
txt = "alisha sajjad"
print(txt.title())

#swapcase()-> upper to lower and lower to upper
txt = "Alisha sajjad"
print(txt.swapcase())

#casefold() -> strong lowercase (for comparison)
txt = "alisha saJJad"
print(txt.casefold())

#2. Search Method
#find() -> return position of word
txt = "alisha sajjad"
print(txt.find("jad"))
print(txt.find("aliy")) #if word not found then return -1

#index() -> return character position
txt = "alisha sajjad"
print(txt.index("a")) # if index not found of character then error occur

#rfind() or rindex() -> position of last occurrence 
txt = "alisha sajjad"
print(txt.rfind("a"))
print(txt.rindex("s"))

## 3. Check Methods (True / False)
#isalpha() -> only characters?
print("hello1".isalpha())

#isdigit() -> only digits?
print("123".isdigit())

#isalnum() -> letters + numbers
print("hello1".isalnum())

#islower(), isupper() -> all lower or upper case ?
#isspace() -> only spaces
print(" ".isspace())

#startswith()
print("hello1".startswith("he"))

#endswith()
print("hello1".endswith("o1"))

## 4. Replace / Split Methods -> Modify Text

# replace()      Replaces text.
# split()        Splits a string into a list.
# rsplit()       Splits a string from the right side.
# splitlines()   Splits a string at newline characters.

text = "Alisha Sajjad"
print(text.split())

## 5. Space Removal Methods

# strip()        Removes spaces from both sides.
# lstrip()       Removes spaces from the left side.
# rstrip()       Removes spaces from the right side.

text = "      Alisha Sajjad     "
print(text.rstrip())

## 6. count() Method -> Counts occurrences of a character or substring
txt = "alisha"
print(txt.count("a"))

## 7. join() Method -> Converts a list into a string
name = ["alisha", "sajjad"]
print("-".join(name))

##8. Center Method 
# Aligns the text in the center

txt = "alisha"
print(txt.center(100))

## 9. zfill() Method -> Adds zeros to the left side
num = "873874"
print(num.zfill(10))

##10. format ()
age = 5
print("age is {}".format(age)) #f-string is recommended

```

# Python Booleans

```python
###Boolean -> True or False
#T and F always Capital
# use for decision 
print(10>9)

print(10 == 9)

a = 200
b = 33
if b > a:
    print("b is greater than a")
else:
    print("b is not greater than a")

#Operator	Meaning	Example	Result
# >	Greater than	10 > 5	True
# <	Less than	5 < 2	False
# ==	Equal	5 == 5	True
# !=	Not Equal	5 != 3	True
# >=	Greater or Equal	5 >= 5	True
# <=	Less or Equal	4 <= 2	False

## bool() Function -> Checks any value and returns either True or False
print(bool("hello"))
print(bool(0))

#False -> False, None, 0, "", {},(), []

## Function Return Boolean 
def myFunc():
    return False
print(myFunc())

## Boolean Function in if
def myFunc1():
    return False
if myFunc1():
    print("Yes")
else:
    print("No")

## isinstance() Function -> Checks whether a variable is of a specific data type
#-> isinstance(variable, datatype)

x = "0"
print(isinstance(x, str))
    

name = input("enter name: ")
if bool(name):
    print("Welcome", name)
else:
    print("Please write your name")

```

# Python Operators

```python
###Operators

print(10+0)

#1. Arithmetic Operators
##Operator	Meaning	Example	Output
#    +	Addition	5 + 3	8
#    -	Subtraction	5 - 3	2
#    *	Multiplication	5 * 3	15
#    /	Division	5 / 2	2.5
#    %	Modulus (Remainder)	5 % 2	1
#   **	Power	2 ** 3	8
#   //	Floor Division	5 // 2	2

x = 18
y = 2

print(x**y)

#2. Assignment Operator

x = 5

# +=
x = 5
x += 3
print(x)

# -=
x -= 1
print(x)

x *= 4
print(x)

x /= 4
print(x)

x %= 3
print(x)

x **= 3
print(x)

x //= 2
print(x)

#Walrus Operator := -> introduce in python 3.8
#Normally

numbers= "8734"
count = len(numbers)
if count > 3:
    print(count)

#Walrus operator
if(count := len(numbers)) > 3:  #assignment + condition in one line
    print(count)


### Ternary Operator  -> short form of if-else

#Normal form
num = 3
if num > 4:
    x = "Weekend"
else:
    x = "Workday"    

print(x)

#short form
x = "Weekend" if num > 4 else "Workday"


num = 6
x = "Week" if num > 5 else "Work"
print(x)
```

# Remaining Python Operators

```python
##Comparison Operator
#compare 2 values
# Operator	Meaning
# ==	Equal
# !=	Not Equal
# >	Greater
# <	Less
# >=	Greater or Equal
# <=	Less or Equal

x = 5
y = 3

print(x == y)
print(x != y)
print(x > y)
print(x < y)

#Chaining

x = 5
print(1 < x < 10)

print(1 < x and x < 10)

##Logical Operators
#  Operator	Meaning
#  and	Both condition True
#  or	Any one or more true
#  not	Result opposite

x = 5
print(x < 5 or x > 10)
print(not(x<3))

### Identity Operators -> Compare Memory Location 
#is
#is not
x = [1,2]
y = [1,2]
print(x == y) ## value compare
print(x is y) ## Memory Compare

z = x
print(x is z)

###Membership operators -> check a value is in collection or not , also use in strings
# use in or not in key word
name = ["ali", "alisha"]
print("al" in name)
print("sajjad" not in name)

text = "alisha sajjad fa23-bcs-373"
print("fa23" in text)

### Bitwise Operators -> work on binary numbers
# Operator	Meaning
# &	AND
# |	OR
# ^	XOR
# ~	NOT
# <<	Left Shift
# >>	Right Shift

print (4 & 5)
print (3 >> 5)
print (6 ^ 7)
print (4 | 8)

###Operator Precedence 
print(100 + 7 * 3) # first multiply then add

#use Parentheses -> high priority
print((200+3) * 4) 

#Order of Precedence  -> PEMDAS Rule
# ()
# **
# * / // %
# + -
# << >>
# &
# ^
# |
# Comparisons (==, >, <, is, in)
# not
# and
# or

```
