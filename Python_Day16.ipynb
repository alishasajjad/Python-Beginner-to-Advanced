### Marker Size

Use the keyword argument markersize or the shorter version, ms to set the size of the markers.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])

plt.plot(y, marker = 'o', ms = 20)
plt.show()
```

### Marker Color

Use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])

plt.plot(y, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

Use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 120])

plt.plot(y, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

```python
## Use both the mec and mfc arguments to color the entire marker:

import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 81, 1, 10])

plt.plot(y, marker = 's', ms = 20, mec = 'r', mfc = 'b')
plt.show()
```

```python
## use Hexadecimal color values:

plt.plot(y, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
```

### Multiple Lines

Plot as many lines as you like by simply adding more plt.plot() functions.

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([3, 8, 1, 10])
x2 = np.array([6, 2, 7, 11])

plt.plot(x1)
plt.plot(x2)

plt.show()
```

Plot many lines by adding the points for the x- and y-axis for each line in the same plt.plot() function.

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([23, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 12])

plt.plot(x1, y1, x2, y2)
plt.show()
```

## Matplotlib Labels and Title

### Create Labels for a Plot

With Pyplot, Use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')

plt.show()
```

### Create a Title for a Plot

With Pyplot, Use the title() function to set a title for the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 10, 105, 110, 15, 120, 125])
y = np.array([240, 250, 60, 20, 280, 290, 300, 310, 320, 330])

plt.plot(x, y, color = 'green', lw= '10.2')

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

### Set Font Properties for Title and Labels

Use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 10, 105, 110, 15, 120, 125])
y = np.array([240, 250, 60, 20, 280, 290, 300, 310, 320, 330])

plt.plot(x, y, color = 'green', lw= '10.2')

font1 = {'family': 'serif', 'color' : 'blue', 'size': 20}
font2 = {'family': 'serif', 'color' : 'darkred', 'size': 12}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.show()
```

### Position the Title

Use the loc parameter in title() to position the title.

Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 105, 120, 125])
y = np.array([240, 250, 160, 20, 280, 290, 300, 31, 320, 330])

plt.plot(x, y, color = 'green', lw= '10.2')

font1 = {'family': 'serif', 'color' : 'blue', 'size': 18}
font2 = {'family': 'serif', 'color' : 'darkred', 'size': 12}

plt.title("Sports Watch Data", fontdict = font1, loc = 'left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.show()
```

## Example

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([10,20,15,25,30])

plt.plot(x, y, marker='o', ms=10, mec='blue', mfc='darkred', ls='--', lw=2, c='green')

plt.title("Student Progress", color = 'darkblue', font = {'family': 'serif','size': 18 })
plt.xlabel("Week")
plt.ylabel("Marks")

plt.show()
```

```python
# Function	            Purpose
# plt.plot(x, y)	Draw a line graph
# plt.show()	    Display the graph
# marker='o'	    Add circle markers
# ms=10	            Marker size
# mec='red'	        Marker edge color
# mfc='blue'	    Marker face color
# ls='--'	        Dashed line
# lw=3	            Line width
# c='green'	        Line color
# plt.xlabel()	    X-axis label
# plt.ylabel()	    Y-axis label
# plt.title()	    Graph title
```
