# Python - Access List Items

# Python - Change List Items

```python
###Python - Change List Items

# List is Changeable (Mutable)

# 1. Change Item Value -> When changing only one item, use its index
# Syntax -> list_name[index] = new_value

student = ["ali", "alisha","ibrahim", "zarmeen", "aiman", "tayyab", "talha"]
student[0] = "Sajjad"
print(student)

# 2. Change Multiple Items -> When changing more than one item, use slicing (Range)
# Syntax -> list_name[start:end] = [new values] # Start is included, end is excluded
student = ["ali", "alisha","ibrahim", "zarmeen", "aiman", "tayyab", "talha"]
student[2:5] = ["Ahmad", "Ahsan", "Hadia"]
print(student)

##When inserting more items
student = ["ali", "alisha","ibrahim", "zarmeen", "aiman", "tayyab", "talha"]
student[1:2] = ["Zarmeen", "IBRAHIM"]
print(student)

##When inserting fewer items
student = ["ali", "alisha","ibrahim", "zarmeen", "aiman", "tayyab", "talha"]
student[1:3] = ["Zarmeen"]
print(student)

###INSERT ITEMS

# When adding a new item at any position in the list without deleting existing items
# Syntax -> list.insert(index, value)
student = ["ali", "alisha","ibrahim", "zarmeen", "aiman", "tayyab", "talha"]
student.insert(0, "Sajjad")
print(student)

###Insert vs Assignment

# Assignment -> Replaces an existing item
student = ["ali", "alisha","ibrahim", "zarmeen", "aiman", "tayyab", "talha"]
student[1] = "alishy"
print(student)

# Insert -> Adds a new item without replacing existing items
student.insert(4, "AIMAN")
print(student)

## =  ⇒ Replace
## insert() ⇒ Insert + Shift
```

# Python - Add List Items

```python
### Python - Add List Items

# Add new items to a list
# 3 common methods
# append(), insert(), extend()

# 1. append()
# append() always adds an item at the end of the list
# Syntax -> list_name.append(value)

student = ["ali", "alisha","ibrahim", "zarmeen", "aiman", "tayyab", "talha"]
student.append("Ahsan")
print(student)

# 2. insert()
# Adds an item at a specific position
# Syntax -> list_name.insert(index, value)

student = ["ali", "alisha","ibrahim", "zarmeen", "aiman", "tayyab", "talha"]
student.insert(2, "Hadia")
print(student)

# 3. extend()
# Used to add all elements of one list into another list
# Syntax -> List1_name.extend(List2_name)

List1 = [1,2,4,5]
List2 = [6,7,8,9,3]
List1.extend(List2)
print(List1)

### append vs extend

#append()
List1 = [1,2,4,5]
list4 = [6,7,8,9,3]
List1.append(list4)
print(List1)

#extend()
List1 = [1,2,4,5]
List2 = [6,7,8,9,3,"alisha"]
List1.extend(List2)
print(List1)

# extend() accepts any iterable, not just lists
# An iterable is an object whose items can be accessed one by one

list1 = ["Alisha", "Sajjad"]
tuple1 = ("Sajjad", "Alisha")

list1.extend(tuple1)
print(list1)

## String
list1 = ["Alisha", "Sajjad"]
list1.extend("gta")
print(list1)

# A string is also an iterable
# Python adds each character separately

# Set Example
list1 = ["alisha"]
set1 = {"alisha1", "sajjad", "ali"}
list1.extend(set1)
print(list1)

# A set is unordered
```

# Python - Remove List Items

```python
###Python - Remove List Items

# 4 main methods

#1. remove()
# Removes an item by its value
# Syntax -> list_name.remove(value)

list1 = ["item1", "item2", "item3"]
list1.remove("item2")
print(list1)

# If duplicate values exist, remove() deletes the first matching value
list1 = ["item1", "item2", "item3", "item2"]
list1.remove("item2")
print(list1)
# remove() -> Removes the first occurrence

#2. pop() -> Removes an item using its index
# Syntax -> list_name.pop(index)

list1 = ["item1", "item2", "item3"]
list1.pop(0)
print(list1)

# If no index is provided, Python automatically removes the last item
list1 = ["item1", "item2", "item3"]
list1.pop()  #pop() -> pop(-1)
print(list1)

# pop() removes and returns the removed item
list1 = ["item1", "item2", "item3"]
x = list1.pop(2)
print(x) # The removed item is stored in x
print(list1)

#3. del -> A keyword used to delete an item by its index
# Syntax -> del list_name[index]

list1 = ["item1", "item2", "item3"]
del list1[1]
print(list1)

# del can also delete the entire list

list1 = ["item1", "item2", "item3"]
del list1
#print(list1)

#4. clear() -> Does not delete the list itself
# It removes all items and returns an empty list
list1 = ["item1", "item2", "item3"]
list1.clear()
print(list1) #Empty list object
```

# Python - Loop Lists

```python
###Python - Loop Lists -> Access all list items one by one

# Loop -> Automatically accesses every item in a collection and performs an operation
student = ["ahsan", "alisha","ibrahim", "hadia", "aiman", "tayyab", "talha", "ahmad"]

#1. Loop through a list (for loop)
# for variable in list:
#       print(variable)

for x in student: #x is a temporary variable 
    print(x)# Prints each item directly

#2. Loop Through Index Numbers
# Used when you need both the index and the item
# Use range() with len()

for i in range(len(student)): #index, item
    #print(student[i])
    print(i, student[i])

#3. Using While Loop -> Traverse a list using a while loop

list1 = ["alisha", "sajjad", "ibrahim", "asif", "ali"]
i = 0
while i < len(list1): # When this condition becomes False, the loop stops
    print(list1[i])
    i = i + 1  # If this line is missing, an infinite loop will run
    
#4. List Comprehension

student = ["ali", "alisha","ibrahim", "zarmeen", "aiman", "tayyab", "talha"]
[print(x) for x in student] #short form

# for -> Short, easy, most common, and recommended
# while -> Used when you need to manage the index manually or loop based on a condition
# list comprehension -> Short syntax, mostly used for creating new lists 

```

# Python-List Comprehension

```python
###Python - List Comprehension

# A shortcut (short syntax) used to create a new list from an existing list

# Without List Comprehension (Normal Method)
student = ["ali", "alisha","ibrahim", "zarmeen", "aiman", "tayyab", "talha"]
newList = []

for x in student:
    if "t" in x:
        newList.append(x)
print(newList)

#With List Comprehension
student = ["ali", "alisha","ibrahim", "zarmeen", "aiman", "tayyab", "talha"]

newList = [x for x in student if "t" in x] # [variable , for loop, condition] #[expression for item in iterable if condition]
# expression -> Defines what will be added to the new list
# for loop -> Iterates through each item
# if condition -> Filters the items
print(newList)

###Condition 
newList = [x for x in student if x != "alisha"]
print(newList)

##Without if -> Copies every item
newList = [x for x in student]
print(newList) #Copy of the Original List

##Iterable means -> An object that can be iterated over using a loop
# List, Tuple, Set, String, range()

# Condition with range()
newList = [x for x in range(10) if x < 5]
print(newList)

# Expression -> Can perform operations, not just return the item
newList = [x.upper() for x in student]
print(newList)

newList = ["hello" for x in student]
print(newList)

##if-else expression 
newList = [x if x != "alisha" else "ali" for x in student]
print(newList)

#Traditional Loop vs List Comprehension
newList1 = []
for x in student:
    if "a" in x:
        newList1.append(x)
        
print(newList1)

newList2 = [x for x in student if "a" in x]
print(newList2)
```

# Python Sort Lists

```python
#1. sort() Method -> Sorts in ascending order

# sort() permanently changes the original list
student = ["ali", "alisha","ibrahim", "zarmeen", "aiman", "tayyab", "talha"]
student.sort()
print(student)

numbers = [98,87,34,0,21,129]
numbers.sort()
print(numbers)

#2. Descending Sort

# Syntax -> list_name.sort(reverse=True)
student = ["ali", "alisha","ibrahim", "zarmeen", "aiman", "tayyab", "talha"]
student.sort(reverse = True)
print(student)

numbers = [98,87,34,0,21,129]
numbers.sort(reverse =True)
print(numbers)

#3. key Function -> Defines a custom sorting rule

# key means -> Python first applies the function to each item,
# then compares the returned values for sorting 
def myfunc(n):
    return abs(n-50)

numbers = [100,50,65,82,23]
numbers.sort(key=myfunc)  # Sort using absolute distance from 50

# Sorts according to the nearest value to 50
# Returned values: 50, 0, 15, 32, 27
# Python compares these returned values and then arranges the original list

print(numbers)

#4. Case-Sensitive Sort
student = ["ali", "Alisha","ibrahim", "Zarmeen", "aiman", "tayyab", "talha"]
student.sort()
print(student)

# According to the ASCII table, uppercase letters (A-Z)
# come before lowercase letters (a-z)

#5. Case-Insensitive Sort -> Converts all strings to lowercase before comparing
student = ["ali", "Alisha","ibrahim", "Zarmeen", "aiman", "tayyab", "talha"]
student.sort(key=str.lower)
print(student)

#6. reverse() -> Reverses the current order of the list
student = ["ali", "Alisha","ibrahim", "Zarmeen", "aiman", "tayyab", "talha"]
student.reverse()
print(student)
```

# Python Match

```python
###PYTHON MATCH

# The match statement is used to perform different actions based on different conditions
# match is used to execute different code based on a single value

# It removes the need for multiple if...elif...else statements

# Syntax
#match expression:

#     case value1:
#        code
#     case value2:
#         code
#     case_:
#         default code


day = 4

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 4:
        print("Thursday")
    case 5: 
        print("sunday")
        
###Match vs if

#if
day = 2

if day==1:
    print("Monday")
elif day==2:
    print("Tuesday")
elif day==3:
    print("Wednesday")


#match
match day:

    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")

##Default Case(_)
#same as else
day = 10

match day:
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _:
        print("default")

# Rule -> The underscore (_) should always be the last case

# match day:
#     case _:    # _ matches every value, so the cases below will never execute
#         print("Hello")
#     case 1:
#         print("Monday")

###Multiple Values

# Use the pipe (|) operator, which means OR

day = 2
match day:
    case 1|2|3|4|5:
        print("WeekDAy")
    case 6|7:
        print("Weekend")

###Guards (if)

# Sometimes checking only the value is not enough
# When an additional condition is required, use if (guard)

month = 5
day = 3

match day:
    case 1|2|3|4|5 if month == 4:
        print("April")
    case 1|2|3|4|5 if month == 5:
        print("May")
    case _:
        ("no match")
        

# Use match when checking multiple possible values of the same variable
# Examples: day, month, grade, menu, option, command

option = "c"

match option:
    case "a":
        print("Add")
    case "b":
        print("Delete")
    case "c":
        print("Update")
    case 4:
        print("Exit")
        

# Use if when comparisons are more complex
# Examples: marks, age, salary, etc.

# Feature              Purpose

# match                Compares the value of a single expression
# case                 Defines a block for each possible value
# _                    Default case (same as else)
# |                    Multiple values (OR)
# if (Guard)           Adds an extra condition to a case

```

# Python *args and **kwargs

```python
#*agrs -> collect multiple positional arguments into a tuple

#*kwargs -> collect multiple keyword arguments into a dictionary

#Normal Function
def student(name, age):
    print(name)
    print(age)

student("Alisha", 22) #function only for 2 arguments -> name and age
#if 3 arguments pass -> error
#student("Alisha",22,"CS")  -> error -> takes 2 positional arguments but 3 were given

#suppose we don't know -> user -> how many arguments will pass  -> then use *args and **kwargs

#*args
# * means -> all extra positional arguments will be collected into a tuple

#syntax -> def function_name(*args):

def student1(*names):
    print(names) #names is a tuple
    print(type(names))

student1("Alisha") 
student1("Alisha", "ali", "ahmad") #function call with 3 arguments 

###Access Values in *args
def student2(*names):
    print(names[0])
    print(names[1])

student2("Alisha", "Ali", "34") #function call with 2 arguments -> output Alisha Ali

#Loop 
def student3(*names):
    for name in names:
        print(name)

student3("Alisha", "Ali", "34") #function call with 3 arguments -> output Alisha Ali 34

#Example

def add(*numbers):
    total = 0 #initialize total to 0
    
    for i in numbers:
        total += i #add each number to total
        
    print("Total:", total) #print total

add(1,2,3,4,5) #function call with 5 arguments  
add(10,20,30) #function call with 3 arguments  

## Regular Parameter + *args

def greet(message, *names):
    print(message)
    print(names)
    
greet("Welcome", "ali", "alisha", "alisha1", "45") #function call with 1 regular parameter and 4 *args

#Loop 
def greet1(message, *names):
    
    for i in names:
        print(message, i)
    
greet1("Welcome", "ali", "alisha", "alisha1", "45")

### **kwargs

#suppose we want user data 
# sometime name, or name,age, etc 
#parameters not fixed -> then use **kwargs

def student12(**data):
#** means -> all extra keyword arguments will be collected into a dictionary
    print(data) #data is a dictionary
    print(type(data))

#student12("alisha") #error -> takes 0 positional arguments but 1 was given

student12(name="alisha") #function call with 1 keyword argument
student12(name="alisha", age=22) #function call with 2 keyword arguments
student12(name="alisha", age=22, status = "login")

###Access Values in **kwargs
def student13(**data):
    print(data["name"]) #access value by key

student13(name="alisha", age=22) #function call with 2 keyword arguments

###Loop 
def student14(**data):
    for key,value in data.items():
        print(key, value)

student14(name="alisha", age=22, status = "login") #function call with 3 keyword arguments

###Regular Parameter + **kwargs
def student15(message, **data):
    print(message)
    print(data)
    
student15("Welcome", name="alisha", age=22, status = "login") #function call with 1 regular parameter and 3 **kwargs

### *args + **kwargs
def info(title, *emails, **data):
    print(title)
    print(emails)
    print(data)

info("Student Info", "alisha@example.com", "ali@example.com", name="alisha", age=22, status="login")

#Order -> Normal Parameter -> *args -> **kwargs

### Unpacking (*)

#previous we collect multiple arguments into a tuple or dictionary using *args and **kwargs
#now we can unpack a tuple or dictionary into multiple arguments using *

#now we expand values

def add(x,y,z):
    print(x+y+z)

#normally -> add(1,2,3) #function call with 3 arguments
#but values is in a tuple -> then use unpacking
numbers1 = (1,2,3) #tuple
add(*numbers1) #unpacking the tuple into individual arguments
print(numbers1) #output [1,2,3]

print(type(numbers1)) #output <class 'tuple'>
print(numbers1[0]) #output 1

###Unpacking (**)
#similarly we can unpack a dictionary into multiple keyword arguments using **

person = {"name": "alisha", "age": 22, "status": "login"} #dictionary
def student16(name, age, status):
    print(name)
    print(age)
    print(status)

student16(**person) #unpacking the dictionary into individual keyword arguments

print(person["age"])

#*args = Collect multiple positional arguments → Tuple
#**kwargs = Collect multiple keyword arguments → Dictionary
```

# Python Decorators

```python
#Python Decorators

# A decorator adds extra functionality to a function
# without changing the original function's code

#Without Decorator
def greet_new():
    return "alisha"

print(greet_new())
# Now we want the output in uppercase

# Simple Method
def greet_new1():
    return "Hello"

print(greet_new1().upper())

# But if there are 100 functions,
# should we write .upper() for every function?

# To solve this problem, use a decorator

#Basic Decorator

# def function_name(func):

def changecase(func):
    def myinner():
        return func().upper()
    return myinner

# changecase(greet)
# func = greet

# Inner function -> A function defined inside another function

# return func().upper() -> First execute the original function,
# then apply upper()

# return myinner -> Return the function itself

# return myinner() -> Execute the function immediately

###Decorator Usage

@changecase
def greet():
    return "hello123"

greet()

#Python Internally 
def greet1():
    return "hello_new"

greet22 = changecase(greet1)
print(greet22())

###Example
def changecase1(func1):

    def myinner1():
        return func1().lower()

    return myinner1

@changecase1
def greet23():
    return "ALISHA SAJJAD"

print(greet23())

#Multiple Functions
@changecase
def greeter():
    return "Hello"
greeter()

@changecase
def city():
    return "Lahore"
city()

###Function with Arguments

def change1(func_new):
    def inner_new(name):
        return func_new(name).upper()
    return inner_new

@change1
def new1(name):
    return "Hello " + name

print(new1("zarmeen"))

# Problem -> If a function has two or more arguments,
# the wrapper must be changed every time

# Solution -> *args and **kwargs

### *args and **kwargs

# Universal wrapper for decorators

# def innerfunc1(*args, **kwargs):
#     return new_func1(*args, **kwargs).upper()

###Example
def changing(newfunc1):
    def inner11(*email, **data):
        return newfunc1(*email, **data).upper()
    return inner11

@changing
def data(email, data):
    return "hi " + email + data

print(data("@gmail.com ", data = "alisha" ))

#Decorators with Arguments

# Suppose you want to choose uppercase or lowercase

# @changecase(1)
# or
# @changecase(2)

# For this, create one more wrapper

#Decorator Argument
#↓
#Decorator
#↓
#Wrapper
#↓
#Original Function

##Example
def alisha1(choice):
    def decorator(func):
        def wrapper():
            if choice == 1:
                return func().upper()
            
            else:
                return func().lower()
        return wrapper
    return decorator

#use
@alisha1(1)
def string11():
    return "alisha_sajjad"

print(string11())

@alisha1(2)
def string2():
    return "hello1233we"

print(string2())

#Multiple Decorators

# If multiple decorators are applied,
# the decorator closest to the function runs first

def addgreeting(func):
    def wrapper():
        return "Hello " + func()
    return wrapper

def changecase(func):
    def wrapper():
        return func().upper()
    return wrapper

@changecase
@addgreeting
def greet():
    return "Alis"

print(greet())

# Rule: Decorators execute from bottom to top.
# The decorator closest to the function executes first.

###Function Metadata

# Every function has metadata

def new1():
    pass

print(new1.__name__)

# After applying a decorator
@changecase
def new2():
    pass
print(new2.__name__)

# Why does it print wrapper instead of new2?
# Because the decorator replaces the original function
# with the wrapper function.

# functools.wraps
import functools
def case1(func_new):
    @functools.wraps(func_new)
    
    def inner_new():
        return func_new().upper()
    
    return inner_new

@case1
def greett():
    return "abc"

print(greett.__name__) # The original function name is preserved.

#Original Function
#   ↓
#Decorator Receive
#   ↓
#Wrapper Function
#   ↓ 
#Extra Work
#   ↓
#Original Function
#   ↓
#Modified Result
#   ↓
#Return

##Examples
#Example 1: Login Check

def login_required(login_func):
    def inner_func(gmail):
        return login_func(gmail).lower()
    return inner_func

@login_required
def profile(gmail):
    return gmail + "@gmail.com"

print(profile("ALiSha12345"))


#Example 2: Timing
#@timer
#def train_model():

#Example 3: Permission
#@admin_only
#def delete_user():
```

# Python Recursion

```python
#Recursion

# Recursion means a function calls itself

def hello():
    print("hello")
    hello()  # The function calls itself -> infinite recursion

#hello()

# Base Case

# Every recursive function must have a stopping condition.
# This stopping condition is called the base case.

def hello1(n):
    if n == 0:
        print("Done")
    else:
        print(n)
        hello1(n-1)
hello1(8)

# Factorial -> 5! = 5 × 4 × 3 × 2 × 1

# Recursive solution
def factorial(n):
    if n == 1:
        return 1
    return n*factorial(n-1)

print(factorial(5))

# Fibonacci

# Rule -> The next number is the sum of the previous two numbers

# 0, 1, 1, 2, 3, 5, 8, 13, 21

def fibonacci(n):
    if n<=1:
        return n
    return fibonacci(n-1)+fibonacci(n-2)

print(fibonacci(5))

#fib(3) = fib(2) + fib(1)
#       = 1 + 1

#Fibonacci sequence
# n      Result
# 0  →   0
# 1  →   1
# 2  →   1
# 3  →   2
# 4  →   3
# 5  →   5
# 6  →   8
# 7  →   13

##Recursion Limit

# By default, Python allows approximately 1000 recursive calls
import sys
print(sys.getrecursionlimit()) 
#set limit

sys.setrecursionlimit(4000)
print(sys.getrecursionlimit())
```

# Python Generators

```python
#Normal Function 
def number11():
    return [1,2,3]

# The entire list is created and stored in memory

#Generator
def num1():
    yield 1
    yield 2
    yield 3

# No list is created.

# Values are generated one by one.

# This makes generators memory-efficient.

##yield vs return

# return -> Ends the function immediately
def test1():
    return 5  # Function ends here
    return 10  # This line is never executed

print(test1()) #output -> 5

# yield -> Pauses the function instead of ending it
def test2():
    yield 1
    yield 2
    yield 3

for i in test2():
    print(i)

#next()

def test3():
    yield "alisha"
    yield "zarmeen"
    yield "Ms.samra"

x = test3()
print(next(x))
print(next(x))
print(next(x))
# print(next(x))  # ERROR -> StopIteration (No more values are available)

###Memory Saving

# Suppose we need 10 million numbers

# List

# Expression -> [x * x for x in range(5)]
number123 = [x for x in range(10000000)] # The entire list is stored in RAM 

# Generator

# Expression -> (x * x for x in range(5))
# It does not create a list; it returns a generator object
number456 = (x for x in range(10000000))
# Only one value is generated at a time.

# This is why generators are very memory-efficient.

# Convert to a list
list(number456)


#yield
# ↓
#Pause
# ↓
#Resume
# ↓
#Memory save
# ↓
#Returns an iterator

#Lambda vs Normal Function

# Normal Function           Lambda
# def is used               lambda is used
# Multiple statements       Only one expression
# Has a name                Anonymous
# Suitable for large code   Suitable for small functions

#Generator vs Normal Function

# Normal Function                 Generator
# return                          yield
# Returns all values at once      Returns values one by one
# Function ends                   Function pauses
# Uses more memory                Uses less memory

#Recursion vs Loop

# Loop                            Recursion
# Uses for/while                  A function calls itself
# Faster                          Can be slower in some cases
# Easier                          Slightly more difficult
# Does not use the call stack     Uses the function call stack
```
