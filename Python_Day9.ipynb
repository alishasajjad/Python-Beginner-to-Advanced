## NumPy Random Numbers

```python
## A random number is a number that cannot be predicted

#If you toss a coin, you cannot predict whether it will be Head or Tail.
#Similarly, random numbers are unpredictable
```

### Pseudo Random vs True Random

```python
## Pseudo Random

# Computers generate random numbers using an algorithm (program)
# Since the computer follows instructions, these numbers are not truly random

# They are called Pseudo Random Numbers

from numpy import random

print(random.randint(100))
print(random.randint(100))
print(random.randint(100))
print(random.randint(100))

# Output (changes every run)

# Although the output looks random, it is generated by an algorithm
```

```python
## True Random

# True random numbers come from outside sources such as:

#Mouse movement
#Keyboard typing
#Weather sensors
#Network activity

# These are mostly used in:

#Encryption
#Security
#Password generation

#For normal programming and data science, Pseudo Random Numbers are sufficient
```

### random.randint()

```python
# random.randint()

# Used to generate random integers

#Syntax ->   random.randint(max)

from numpy import random

x = random.randint(121)
y = random.randint(121)
z = random.randint(121)
print(x)
print(y)
print(z)

# Output : The number is between 0 and 120
```

```python
## Generate Multiple Random Integers

x = random.randint(100, size = 5)
print(x)

# 100 → numbers from 0–99
# size=5 → generate 5 numbers
```

```python
## Generate 2D Random Integer Array

x = random.randint(100, size = (3,5))
# (3,5) -> 3 rows and 5 columns

print(x)
```

### random.rand()

```python
## random.rand()

# Used to generate random decimal (float) numbers

# Range: 0 ≤ number < 1

x = random.rand()
print(x)
```

```python
## Multiple Float Numbers

x = random.rand(4)
print(x)
```

```python
## 2D Float Array

x = random.rand(3,4)
# (3,4) -> 3 rows, 4 columns

print(x)
```

### random.choice()

```python
## random.choice()

# Used to randomly select values from an existing array

x = random.choice([3,4,2,1])
y = random.choice([3,4,2,1])
print(x)
print(y)
```

```python
## Generate Multiple Random Choices

x = random.choice([12,4,2,4,623,1,0], size = 3)
print(x)
```

```python
## Generate 2D Random Array

x = random.choice([12,3,2,6,0,9,234,876,4523,21,3], size = (3,4))
print(x)
```

## Random Data Distribution

```python
## Random Data Distribution

# A data distribution tells: 
# Which values are possible
# How likely each value is to appear

#Example

#Suppose we have four values:

# 3
# 5
# 7
# 9

# want

# 3 → 10%
# 5 → 30%
# 7 → 60%
# 9 → 0%

# Instead of giving every value the same chance, we assign different probabilities
```

### Probability (p)

```python
## Probability (p)
num = [3,5,7,9]
p = [0.1,0.3,0.6,0]

#Means

# Value	Probability
# 3	    0.1 (10%)
# 5	    0.3 (30%)
# 7	    0.6 (60%)
# 9	    0.0 (0%)

# 0.1 + 0.3 + 0.6 + 0 = 1
# The probabilities must always add up to 1
```

```python
## Example

from numpy import random

x = random.choice(
    [3,5,7,9],
    p = [0.1,0.3,0.6,0],
    size = 20
)

print(x)

# 7 appears most often
# 3 appears rarely
# 9 never appears
```

```python
## 2D Probability Distribution

from numpy import random

x = random.choice(
    [3,5,7,9],
    p = [0.1,0.3,0.6,0],
    size = (3,5)
)

print(x)
```

## Random Permutations

```python
## Random Permutations

# A permutation means arranging the same elements in a different order

## Example

# Original:

# [1 2 3]

# Possible permutations:

# [3 2 1]
# [2 1 3]
# [1 3 2]

# The values remain the same; only their order changes
```

### shuffle()

```python
## shuffle()

# Shuffle changes the original array

from numpy import random
import numpy as np

arr = np.array([1,2,4,231,78,45,0])

random.shuffle(arr)

print(arr)

# The original array itself becomes:
# [ 45   0   4  78   2   1 231]

#So shuffle() modifies the existing array in-place

# The original array has changed
```

### permutation()

```python
## permutation()

# permutation() also rearranges the elements, but does not change the original array

import numpy as np
from numpy import random

arr = np.array([1,2,367,89,32,0,211])

new_arr = random.permutation(arr)
print(new_arr)

print(arr)

# The original array remains unchanged
```

```python
arr = np.array([1,2,3,4,5,0])

random.shuffle(arr)

print(arr)

# Original array has changed
```

```python
arr = np.array([1,2,3,4,5,0])

new_arr = random.permutation(arr)

print(new_arr)
print(arr)

# Original array is unchanged
```

```python
# Function	            Purpose
# random.randint()	Generate random integers
# random.rand()	    Generate random float numbers (0–1)
# random.choice()	    Pick random values from an array
# p=	                Assign probability to each value
# shuffle()	        Rearrange elements and modify the original array
# permutation()	    Return a shuffled copy without changing the original array
```
